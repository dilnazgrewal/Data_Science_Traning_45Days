# *Data Cleaning*

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Project\\Superstore_Messy.csv")
```

```python
df.info()
```

```python
# ---- COMPLETE DATA QUALITY SUMMARY ----

print("📋 DATA QUALITY SUMMARY")
print("=" * 60)

print("\n1. MISSING VALUES PER COLUMN:")
print(df.isnull().sum())
print(f"\n   Total missing values: {df.isnull().sum().sum()}")

print(f"\n2. DUPLICATE ROWS: {df.duplicated().sum()}")

print("\n3. DATA TYPES:")
print(df.dtypes)

print(f"\n4. UNIQUE PRODUCTS:  {df['Product Name'].dropna().unique()}")
print(f"   UNIQUE CATEGORIES: {df['Category'].dropna().unique()}")
print(f"   UNIQUE CATEGORIES: {df['Sub-Category'].dropna().unique()}")
```

```python
print("Rows with missing values: ")
missing = df[df.isnull().any(axis=1)]
display(missing)
print(f"\n→ {len(missing)} rows have at least one missing value")
```

```python
print("---- MISSING VALUE SUMMARY TABLE ----")

missing_count = df.isnull().sum()                        
missing_percent = (missing_count / len(df)) * 100        

missing_summary = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing %': missing_percent.round(2)              
})

print("MISSING VALUES — COUNT & PERCENTAGE")
print(missing_summary)
```

```python
df_dropped = df.dropna()

print("BEFORE dropna():")
print(f"Rows: {len(df)}")

print("\nAFTER dropna():")
print(f"  Rows: {len(df_dropped)}")
print(f"  Rows removed: {len(df) - len(df_dropped)}")
print(f"  Data lost: {round(((len(df) - len(df_dropped)) / len(df) * 100),2)}%")
```

```python
df_dropped.isnull().sum()
```

```python
df_profit = df.copy()
median_profit  = df_profit["Profit"].median()
print("Before fillna() missing values:", df_profit["Profit"].isnull().sum().sum())
# print(df_profit[df_profit["Profit"].isnull()][['Order ID', 'Product Name', 'Profit']])
df_profit["Profit"] = df_profit["Profit"].fillna(median_profit)
print("\nAfter fillna() missing values:", df_profit["Profit"].isnull().sum().sum())
```

```python
df_no_duplicates = df.drop_duplicates()

print("Before Duplicates Removal")
print(f"{df.duplicated().sum()} were found to be duplicated")
print("\nAfter Duplicates Removal")
print(f"{df_no_duplicates.duplicated().sum()} were found to be duplicated")
```

```python
print("------ Outliers Summary (Profit) ------")
Q1 = df["Profit"].quantile(0.25)
Q3 = df["Profit"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = round(Q1 - 1.5 * IQR, 2)
upper_bound = round(Q3 + 1.5 * IQR, 2)

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
outliers_profit = df[(df["Profit"] < lower_bound) | (df["Profit"] > upper_bound)]
print(f"\n→ Found {len(outliers_profit)} outliers in Profit")
if len(outliers_profit) > 0:
    print("\nSample of outliers_profit:")
    display(outliers_profit[['Order ID', 'Product Name', 'Profit']].head(10))

df_no_outliers = df.copy()

print("BEFORE CAPPING:")
print(f"Max Profit: {df_no_outliers['Profit'].max()}")

df_no_outliers['Profit'] = df_no_outliers['Profit'].clip(lower=lower_bound, upper=upper_bound)

print("\nAFTER CAPPING:")
print(f"Max Profit: {df_no_outliers['Profit'].max()}")

remaining_outliers = df_no_outliers[(df_no_outliers['Profit'] < lower_bound) | (df_no_outliers['Profit'] > upper_bound)]
print(f"\n✅ Remaining outliers: {len(remaining_outliers)}")

display(df_no_outliers[['Order ID', 'Product Name', 'Profit']])

print("------ Outliers Summary (Sales) ------")
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = round(Q1 - 1.5 * IQR, 2)
upper_bound = round(Q3 + 1.5 * IQR, 2)

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
outliers_profit = df[(df["Sales"] < lower_bound) | (df["Sales"] > upper_bound)]
print(f"\n→ Found {len(outliers_profit)} outliers in Sales")
if len(outliers_profit) > 0:
    print("\nSample of outliers_profit:")
    display(outliers_profit[['Order ID', 'Product Name', 'Sales']].head(10))

print("BEFORE CAPPING:")
print(f"Max Sales: {df_no_outliers['Sales'].max()}")

df_no_outliers['Sales'] = df_no_outliers['Sales'].clip(lower=lower_bound, upper=upper_bound)

print("\nAFTER CAPPING:")
print(f"Max Sales: {df_no_outliers['Sales'].max()}")

remaining_outliers = df_no_outliers[(df_no_outliers['Sales'] < lower_bound) | (df_no_outliers['Sales'] > upper_bound)]
print(f"\n✅ Remaining outliers: {len(remaining_outliers)}")

display(df_no_outliers[['Order ID', 'Product Name', 'Sales']])
```
