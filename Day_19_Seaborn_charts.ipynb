## 1. What is Seaborn?

**Seaborn** is a Python data visualization library built **on top of Matplotlib**.

Think of it this way:
- **Matplotlib** = manual car (powerful, full control, but more effort)
- **Seaborn** = automatic car (easier to drive, looks better by default, but Matplotlib runs under the hood)

### Key benefits of Seaborn:
- 🎨 **Better default styling** — charts look cleaner without extra work
- 📊 **Works directly with DataFrames** — just pass column names
- 🔥 **Statistical charts built-in** — heatmaps, boxplots, etc. are easier
- ⚡ **Less code** for the same chart

---

## 2. Matplotlib vs Seaborn — Quick Comparison

| Feature | Matplotlib | Seaborn |
|---|---|---|
| **Style** | Basic look by default | Beautiful look by default |
| **Code** | More code needed | Less code needed |
| **Data input** | Uses arrays/lists | Works directly with DataFrames |
| **Best for** | Customization & fine control | Quick EDA & statistical charts |
| **Heatmaps** | Complex to create | One line of code |
| **Relationship** | Foundation library | Built on top of Matplotlib |

**Bottom line:** Use **Seaborn for quick, clean EDA charts**. Switch to **Matplotlib when you need fine-grained control**. In real projects, you'll use both together!

## 2.5 Seaborn Key Attributes & Parameters — Quick Reference Guide 🔑

Before we start coding, let’s understand the **common attributes (parameters)** you will see in Seaborn functions throughout this notebook. Knowing these upfront will help you read and customize every chart confidently.

---

### 🎨 `palette` — Color Scheme for the Chart

Controls the **set of colors** used for bars, dots, or boxes in the chart.

| Value | Effect |
|---|---|
| `'Set2'` | Soft, pastel-like qualitative colors (good for categories) |
| `'pastel'` | Very light, gentle colors |
| `'viridis'` | Perceptually uniform gradient from purple → green → yellow |
| `'Blues_d'` | Dark-to-light blue sequential palette |
| `'magma'` | Dark purple → orange → yellow gradient |
| `'coolwarm'` | Diverging palette: blue (negative) ↔ red (positive) — great for heatmaps |
| `['#3498db', '#e74c3c']` | A **custom list of hex colors** — you define exact colors per category |

**How it affects the graph:** Changes the fill colors of bars, boxes, or data points. Different palettes convey different moods — use sequential palettes for ordered data and qualitative palettes for categorical data.

---

### 🌈 `hue` — Color By a Category Column

Splits the data **by a categorical column** and assigns a different color to each group.

| Value | Effect |
|---|---|
| `'Category'` | Colors each bar/dot/box differently based on the `Category` column |
| `'Product'` | Colors each element differently based on the `Product` column |
| Any categorical column name | Splits and colors by that column |

**How it affects the graph:** Adds a **legend** automatically and colors each group differently. In histograms, bars can be stacked or placed side by side. In scatter plots, each group gets a unique color.

---

### 🔲 `edgecolor` — Border Color of Bars/Dots

Adds a **border (outline)** to each bar, box, or data point.

| Value | Effect |
|---|---|
| `'black'` | Adds a black border — makes elements stand out against each other |
| `'white'` | Adds a white border — creates visual separation on dark backgrounds |
| Any color string or hex code | Uses that color as the border |

**How it affects the graph:** Makes individual bars, dots, or boxes **visually distinct** and easier to read. Without it, adjacent elements may blend together.

---

### 📊 `order` — Custom Sorting of Categories

Controls the **order** in which categories appear on the x-axis.

| Value | Effect |
|---|---|
| `['Laptop', 'Smartphone', 'Tablet', ...]` | Shows categories in this **exact order** |
| `df['Product'].value_counts().index` | Orders by **frequency** (most to least) |
| `df.groupby('Product')['Revenue'].mean().sort_values(ascending=False).index` | Orders by **average value** (highest first) |

**How it affects the graph:** Without `order`, Seaborn uses the natural order from the data. With `order`, you can rank categories from highest to lowest (or any custom sequence) for clearer storytelling.

---

### 📏 `errorbar` — Error Bars on Barplots

Controls what the **thin line (error bar) on top of each bar** represents.

| Value | Effect |
|---|---|
| `'sd'` | Shows **standard deviation** — how spread out the values are |
| `None` | **Removes** error bars entirely |

**How it affects the graph:** A longer error bar = more variation/uncertainty. `'sd'` is often preferred in EDA because it shows the actual spread of data, while `'ci'` is better for statistical inference.

---

### 📉 `bins` — Number of Bins in Histograms

Controls how many **groups (bins)** the data is divided into in a histogram.

| Value | Effect |
|---|---|
| `10` (default) | Divides the range into 10 equal-width bins |
| `12` | Divides into 12 bins — slightly finer detail |
| `20` | More bins — shows more granular distribution |
| `5` | Fewer bins — shows broader patterns |

**How it affects the graph:** Too few bins = you miss patterns. Too many bins = chart looks noisy. A good rule of thumb is to start with 10–15 and adjust.

---

### 〰️ `kde` — Kernel Density Estimate Curve

Overlays a **smooth curve** on top of the histogram to show the distribution shape.

| Value | Effect |
|---|---|
| `True` | Draws a smooth density curve over the histogram |
| `False` | Shows only the bars, no curve |

**How it affects the graph:** When `True`, adds a flowing line that shows the overall \"shape\" of the data — useful for seeing if data is skewed, bimodal, or normally distributed.


---

### ✦ `style` — Marker Shape by Category (Scatterplots)

Assigns a **different marker shape** (circle, square, triangle, etc.) to each category in a scatter plot.

| Value | Effect |
|---|---|
| `'Category'` | Each category gets a unique marker shape |
| `'Product'` | Each product gets a unique marker shape |
| Any categorical column | Assigns shapes per group |

**How it affects the graph:** Especially useful for **colorblind-friendly** charts — even without color, you can tell groups apart by shape. Often used together with `hue`.

---

### 💧 `alpha` — Transparency of Elements

Controls how **transparent or opaque** bars, dots, or boxes are.

| Value | Effect |
|---|---|
| `1.0` | Fully opaque (no transparency) |
| `0.7` | Slightly transparent — good for overlapping dots |
| `0.3` | Very transparent — useful when many points overlap |
| `0.0` | Completely invisible |

**How it affects the graph:** Lower alpha makes overlapping data points visible. In scatter plots with many dots, `alpha=0.5–0.7` prevents a cluttered look and reveals density patterns.

---

### ⚫ `s` — Size of Dots in Scatterplots

Controls the **size of each data point** in a scatter plot.

| Value | Effect |
|---|---|
| `20` (default) | Small dots |
| `80` | Medium dots — easy to see |
| `100` | Larger dots — more prominent |
| `200+` | Very large — use sparingly |

**How it affects the graph:** Larger values make dots easier to see, but too large can cause overlapping. Adjust based on how many data points you have.

---

### 🔢 `annot` — Show Numbers on Heatmap Cells

Whether to display the **actual correlation values** inside each heatmap cell.

| Value | Effect |
|---|---|
| `True` | Shows numbers (e.g., 0.85, -0.23) inside each cell |
| `False` | Shows only colors, no numbers |

**How it affects the graph:** `True` is almost always preferred — it lets you read exact values without guessing from color intensity.

---

### 🔡 `fmt` — Number Format in Heatmap Annotations

Controls how the numbers inside heatmap cells are **formatted**.

| Value | Effect |
|---|---|
| `'.2f'` | Shows 2 decimal places (e.g., 0.85) |
| `'.1f'` | Shows 1 decimal place (e.g., 0.9) |
| `'.0f'` | Shows no decimals (e.g., 1) |
| `'d'` | Integer format |

**How it affects the graph:** `'.2f'` gives a good balance of precision and readability for correlation values.

---

### 🗺️ `cmap` — Colormap for Heatmaps

Sets the **color gradient** used in heatmaps.

| Value | Effect |
|---|---|
| `'coolwarm'` | Blue (negative) → white (zero) → red (positive) — best for correlation |
| `'viridis'` | Purple → green → yellow sequential gradient |
| `'YlGnBu'` | Yellow → green → blue |
| `'RdYlGn'` | Red → yellow → green diverging |

**How it affects the graph:** For correlation heatmaps, use a **diverging** colormap (like `'coolwarm'`) so negative and positive correlations have distinct colors.

---

### Other Heatmap Parameters

| Parameter | What It Does | Common Value |
|---|---|---|
| `linewidths` | Width of grid lines **between** heatmap cells | `1` — adds clean separation |
| `linecolor` | Color of those grid lines | `'white'` — looks clean on colored cells |
| `square` | Makes each cell **square-shaped** (equal width & height) | `True` — neat, symmetric look 

---

### 🖼️ `ax` — Target Subplot Axis

When creating **multiple charts in one figure** (subplots), `ax` tells Seaborn **which subplot** to draw on.

| Value | Effect |
|---|---|
| `axes[0]` | Draws on the **first** subplot |
| `axes[1]` | Draws on the **second** subplot |
| Not specified | Draws on the current/default figure |

**How it affects the graph:** Essential for side-by-side chart comparisons. Without `ax`, all charts go on the same figure.

---

### 🎛️ Global Style Settings (Used Once at Setup)

| Function | What It Does | Values Used in This Notebook |
|---|---|---|
| `sns.set_style()` | Sets the **background grid style** for all charts | `'whitegrid'` — clean white background with grid lines. Other options: `'darkgrid'`, `'white'`, `'dark'`, `'ticks'` |
| `sns.set_palette()` | Sets the **default color palette** for all charts | `'Set2'` — soft qualitative colors. Can be overridden per chart using `palette` |

## 5. Countplot — Counting Categories

**What it does:** Counts how many times each category appears.  
**When to use:** When you want to know **"How many of each?"**  
**Business question:** *"How many transactions does each product have?"*

Think of it as a **visual version of `value_counts()`**.

```python
# ============================================
# COUNTPLOT: Visualize the same thing!
# ============================================

plt.figure(figsize=(8, 5))

# sns.countplot automatically counts and plots
# x = column name, data = DataFrame
sns.countplot(x='Product', data=df,order= df["Product"].value_counts().index,
              edgecolor='black')

# order= sorts bars by count (highest first)

plt.xlabel('Product', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.title('Transaction Count by Product', fontsize=14)

plt.tight_layout()
plt.show()
```

### 📝 How to Read This Chart

- Each bar represents a product.
- The height shows **how many transactions** (not revenue!) each product has.
- This tells us which products are **sold most frequently**.

### 💡 Sample Business Insight

> *"Laptop has the highest number of transactions, making it the most frequently purchased product. However, high transaction count doesn't always mean high revenue — we need to check revenue separately."*

```python
# ============================================
# COUNTPLOT #2: Transactions by Category
# ============================================

plt.figure(figsize=(6, 5))

sns.countplot(x='Category', data=df, edgecolor='black',
              palette=['#3498db', '#e74c3c'])

plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.title('Transaction Count by Category', fontsize=14)

plt.tight_layout()
plt.show()       
```

```python
# ============================================
# COUNTPLOT #3: Count of Revenue Tiers
# ============================================

plt.figure(figsize=(6, 5))

sns.countplot(x='Revenue_Tier', data=df, order=['High', 'Medium', 'Low'],
              palette="coolwarm", edgecolor='black')

plt.xlabel('Revenue Tier', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.title('How Many Transactions in Each Revenue Tier?', fontsize=14)

plt.tight_layout()
plt.show()
```

### 💡 Sample Business Insight

> *"The majority of transactions fall in the Low and Medium revenue tiers. High-revenue transactions are fewer but contribute significantly to total revenue. The business could create bundled offers to push more Medium-tier transactions into the High tier."*

## 6. Barplot — Comparing Averages Across Categories

**What it does:** Shows the **average (or other aggregate)** of a numeric column across categories.  
**When to use:** When you want to compare **"Which category has higher/lower average?"**  
**Business question:** *"What is the average revenue per category?"*

**Countplot vs Barplot:**
| | Countplot | Barplot |
|---|---|---|
| Shows | Count (how many) | Average/sum of a number |
| Input | Just one column | One category + one numeric column |
| Example | How many Laptop transactions? | What is the average Laptop revenue? |

```python
total_rev_by_product=df.groupby("Category")["Revenue"].sum().reset_index()
print(total_rev_by_product)
```

```python
# ============================================
# BARPLOT: Total Revenue by Category
# ============================================

plt.figure(figsize=(6, 5))

# x = category column, y = numeric column, data = DataFrame
# By default, barplot calculates the MEAN and shows confidence interval (black line)
sns.barplot(x='Category', y='Revenue', data=df, edgecolor='black',
            palette='Blues_d', errorbar=None)

# errorbar='sd' shows standard deviation instead of default confidence interval

plt.xlabel('Category', fontsize=12)
plt.ylabel('Average Revenue (₹)', fontsize=12)
plt.title('Average Total by Category', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# ============================================
# BARPLOT: Average Revenue by Category
# ============================================

plt.figure(figsize=(6, 5))

# x = category column, y = numeric column, data = DataFrame
# By default, barplot calculates the MEAN and shows confidence interval (black line)
sns.barplot(x='Category', y='Revenue', data=df, edgecolor='black',
            palette='Blues_d', errorbar='sd')

# errorbar='sd' shows standard deviation instead of default confidence interval

plt.xlabel('Category', fontsize=12)
plt.ylabel('Average Revenue (₹)', fontsize=12)
plt.title('Average Revenue by Category', fontsize=14)

plt.tight_layout()
plt.show()
```

### 📝 How to Read This Chart

- Each bar shows the **average revenue** for that category.
- The **thin black line** on top of each bar shows the **standard deviation** — how spread out the values are.
  - A longer black line = more variation in revenue within that category.
  - A shorter black line = revenue values are more consistent.

### 💡 Sample Business Insight

> *"Electronics has a higher average revenue per transaction than Accessories. The larger spread (standard deviation) in Electronics suggests that it includes both moderate and very high-value transactions."*

```python
# ============================================
# BARPLOT #2: Average Revenue by Product
# ============================================

plt.figure(figsize=(9, 5))

# Order by average revenue (highest first)
product_order = df.groupby('Product')['Revenue'].mean().sort_values(ascending=False).index

sns.barplot(x='Product', y='Revenue', data=df, order=product_order,
            palette='viridis', edgecolor='black', errorbar=None,hue="Category")

plt.xlabel('Product', fontsize=12)
plt.ylabel('Average Revenue (₹)', fontsize=12)
plt.title('Average Revenue by Product', fontsize=14)

plt.tight_layout()
plt.show()
```

### 💡 Sample Business Insight

> *"Laptop has the highest average revenue per transaction, followed by Smartphone. This makes sense because these are higher-priced products. Even though Headphones have many transactions, their average revenue is lower due to the lower price point."*

## 7. Histplot — Distribution of Data

**What it does:** Shows how values are **distributed** — like Matplotlib's `plt.hist()` but better looking.  
**When to use:** When you want to understand the **"shape"** of numeric data.  
**Business question:** *"What does the revenue distribution look like?"*

```python
# ============================================
# HISTPLOT: Revenue distribution
# ============================================

plt.figure(figsize=(9, 5))

# Seaborn's histplot — cleaner than Matplotlib's plt.hist()
sns.histplot(data=df, x='Revenue', bins=12, color='steelblue', edgecolor='black',
             kde=True)  # kde=False means no curve overlay (we'll keep it simple)

plt.xlabel('Revenue (₹)', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.title('Distribution of Revenue', fontsize=14)

plt.tight_layout()
plt.show()
```

### 📝 How to Read This Chart

- Same as a regular histogram — **x-axis** shows revenue ranges, **y-axis** shows count.
- Seaborn just makes it look better with minimal effort!

### 💡 Sample Business Insight

> *"Revenue distribution is right-skewed — most transactions generate low-to-moderate revenue, with a few high-value transactions. This is typical in retail where affordable products sell more often."*

```python
# ============================================
# HISTPLOT #2: Revenue by Category (hue)
# ============================================

# The 'hue' parameter splits the histogram by a category column!

plt.figure(figsize=(9, 5))

sns.histplot(data=df, x='Revenue',hue="Category", bins=12,
             edgecolor='black',kde=False,multiple='dodge')

# multiple='stack' stacks the bars; use 'dodge' to place them side by side

plt.xlabel('Revenue (₹)', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.title('Revenue Distribution by Category', fontsize=14)

plt.tight_layout()
plt.show()
```

### 💡 Sample Business Insight

> *"Electronics transactions are spread across a wider revenue range, while Accessories transactions are concentrated in the lower revenue range. This confirms that Electronics drives the high-revenue segment."*

## 8. Boxplot — Outliers and Spread by Category

**What it does:** Same as Matplotlib boxplot, but Seaborn makes it much **easier to compare across categories**.  
**When to use:** When you want to **compare distributions and spot outliers** across groups.  
**Business question:** *"How does revenue vary across products? Are there outliers?"*

```python
# ============================================
# BOXPLOT: Revenue by Product
# ============================================

plt.figure(figsize=(9, 5))

# x = category, y = numeric, data = DataFrame
# Seaborn handles the grouping automatically!
sns.boxplot(x='Product', y='Revenue', data=df, palette='pastel',
            order=['Laptop', 'Smartphone', 'Tablet', 'Smartwatch', 'Headphones'])

plt.xlabel('Product', fontsize=12)
plt.ylabel('Revenue (₹)', fontsize=12)
plt.title('Revenue Distribution by Product — Spot the Outliers!', fontsize=14)

plt.tight_layout()
plt.show()

df.groupby('Product')['Revenue'].quantile([0.25, 0.50, 0.75])
```

### 📝 How to Read This Chart

- Each **box** represents one product's revenue distribution.
- The **line inside** the box = median revenue for that product.
- The **height** of the box = spread of the middle 50% of data.
- **Diamonds/dots** beyond the whiskers = potential outliers.

**Compare products at a glance:**
- Wide box = lots of variation in revenue
- Narrow box = consistent revenue
- Box higher on y-axis = higher revenue

### 💡 Sample Business Insight

> *"Laptop has the highest median revenue but also the widest spread, meaning some Laptop transactions generate significantly more revenue than others (likely due to quantity differences). Headphones have the lowest and most consistent revenue range."*

```python
# ============================================
# BOXPLOT #2: Revenue by Category
# ============================================

plt.figure(figsize=(7, 5))

sns.boxplot(x='Category', y='Revenue', data=df, palette='Set2')

plt.xlabel('Category', fontsize=12)
plt.ylabel('Revenue (₹)', fontsize=12)
plt.title('Revenue Distribution by Category', fontsize=14)

plt.tight_layout()
plt.show()
```
