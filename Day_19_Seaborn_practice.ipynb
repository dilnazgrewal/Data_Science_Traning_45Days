```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')

sns.set_palette('Set2')

print("Libraries imported!")
```

```python
df = pd.read_excel("DataSet_used\\sales_data.xlsx")

print("Raw data shape: ", df.shape)
df.head()
```

```python
plt.figure(figsize = (10,6))

sns.countplot(x = 'Product', data = df, order = df['Product'].value_counts().index, edgecolor = 'black')

plt.xlabel('Product')
plt.ylabel('Number of Transactions')
plt.title('Transaction Count by Product')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,6))

sns.countplot(x = 'Category', data =df, order = df['Category'].value_counts(ascending=True).index, edgecolor = 'black', palette=('lightblue', 'lightgreen'))

plt.xlabel('Category')
plt.ylabel('Number of Transactions')
plt.title('Transaction Count by Category')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (8,6))

sns.countplot(x = 'Revenue_Tier', data =df, order =['Low', 'Medium', 'High'], edgecolor = 'black', palette=('lightblue', 'grey', 'lightgreen'))

plt.xlabel('Revenue Tier', fontsize = 13)
plt.ylabel('Number of Transactions', fontsize = 13)
plt.title('How Many Transactions in Each Revenue Tier?', fontsize = 16)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (8,6))

sns.barplot(x = 'Category', y = 'Revenue', data =df, edgecolor = 'black', palette= 'Blues_d', errorbar= 'sd')

plt.xlabel('Category', fontsize = 13)
plt.ylabel('Average Revenue', fontsize = 13)
plt.title('Average Total by Category', fontsize = 16)

plt.tight_layout()
plt.show()
```

```python
total_rev = df.groupby('Category')['Revenue'].sum().reset_index()
```

```python
plt.figure(figsize=(9, 5))

# Order by average revenue (highest first)
product_order = df.groupby('Product')['Revenue'].mean().sort_values(ascending=False).index

sns.barplot(x='Product', y='Revenue', data=df, order=product_order,
            palette='viridis', edgecolor='black', errorbar=None,hue="Category")

plt.xlabel('Product', fontsize=12)
plt.ylabel('Average Revenue (₹)', fontsize=12)
plt.title('Average Revenue by Product', fontsize=14)

plt.tight_layout()
plt.show()
```
