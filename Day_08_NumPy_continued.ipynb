## 📌 Topic 9: Mathematical / Vectorized Operations

### 🧑‍🏫 Theory

One of NumPy's greatest powers is **vectorized operations** — applying a mathematical operation to **every element** of the array **without using a loop**.

### ➕ Arithmetic Operations

You can use standard operators `+`, `-`, `*`, `/`, `**` (power), `//` (floor division), `%` (modulo) directly on arrays.

```python
arr = np.array([1, 2, 3, 4])

arr + 10   → [11, 12, 13, 14]   # Add 10 to every element
arr * 3    → [ 3,  6,  9, 12]   # Multiply every element by 3
arr ** 2   → [ 1,  4,  9, 16]   # Square every element
```

### ➕ Element-wise Operations Between Two Arrays

When two arrays have the **same shape**, operations happen **element by element**:

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

a + b  → [5, 7, 9]    # 1+4, 2+5, 3+6
a * b  → [4, 10, 18]  # 1*4, 2*5, 3*6
```

### 📐 Built-in Math Functions

NumPy has powerful built-in math functions:

| Function | What it does |
|----------|--------------|
| `np.sqrt(arr)` | Square root of each element |
| `np.abs(arr)` | Absolute value of each element |
| `np.exp(arr)` | e raised to the power of each element |
| `np.log(arr)` | Natural logarithm of each element |
| `np.round(arr, n)` | Round to n decimal places |

````python
# ============================================================
# TOPIC 9 EXAMPLE: Mathematical / Vectorized Operations
# ============================================================

import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print("Original Array:", arr)
print()

# --- Operations with a single number (scalar) ---
print("arr + 10 :", arr + 10)    # Add 10 to each element
print("arr - 1  :", arr - 1)     # Subtract 1 from each
print("arr * 3  :", arr * 3)     # Multiply each by 3
print("arr / 2  :", arr / 2)     # Divide each by 2
print("arr ** 2 :", arr ** 2)    # Square each element
print("arr % 2  :", arr % 2)     # Modulo (remainder when divided by 2)

print()

# --- Element-wise operations between two arrays ---
a = np.array([10, 20, 30, 40])
b = np.array([1,  2,  3,  4])

print("a:", a)
print("b:", b)
print("a + b:", a + b)    # [11, 22, 33, 44]
print("a - b:", a - b)    # [9, 18, 27, 36]
print("a * b:", a * b)    # [10, 40, 90, 160]
print("a / b:", a / b)    # [10.0, 10.0, 10.0, 10.0]
````

````python
# ============================================================
# TOPIC 9 EXAMPLE: Built-in Math Functions
# ============================================================

import numpy as np

arr = np.array([4, 9, 16, 25, 36])
print("Array:", arr)
print()

# Square root
print("np.sqrt(arr)  :", np.sqrt(arr))

# Absolute value
neg_arr = np.array([-5, -3, 0, 2, 7])
print("np.abs(neg_arr):", np.abs(neg_arr))

# Rounding
decimals = np.array([1.234, 5.678, 9.012])
print("np.round(decimals, 1):", np.round(decimals, 1))

print()

# --- Practical Example: Grade Percentage Boost ---
raw_scores = np.array([70, 80, 55, 90, 65])
print("Raw Scores      :", raw_scores)

# Add 5 bonus marks
boosted = raw_scores + 5
print("After +5 bonus  :", boosted)

# Calculate percentage out of 110 (boosted max)
percentage = np.round((boosted / 110) * 100, 2)
print("Percentage (%)  :", percentage)
````

## 📌 Topic 10: Aggregation Functions

### 🧑‍🏫 Theory

**Aggregation** means summarizing a collection of numbers into a **single value**.  
NumPy provides powerful functions to compute statistics over an array.

| Function | What it computes |
|----------|-----------------|
| `np.sum()` | **Sum** of all elements |
| `np.mean()` | **Average (Mean)** of all elements |
| `np.min()` | **Minimum** (smallest) value |
| `np.max()` | **Maximum** (largest) value |
| `np.median()` | **Median** (middle value when sorted) |
| `np.std()` | **Standard Deviation** (spread of data) |
| `np.var()` | **Variance** (square of std deviation) |

---

### 🏫 Quick Definitions

- **Mean**: Add all values, divide by count. Example: Mean of [60, 80, 100] = (60+80+100)/3 = 80
- **Median**: Sort the values, pick the middle one. Example: Median of [60, 80, 100] = 80
- **Std Dev**: How far (on average) each value is from the mean. **Small std = data is close together, Large std = data is spread out**
- **Variance**: Standard Deviation squared. Another way to measure spread.

---

### 🗂️ Aggregation Along an Axis (for 2D arrays)

For 2D arrays, you can compute aggregations along:
- `axis=0` → **Down each column** (row-wise collapse)
- `axis=1` → **Across each row** (column-wise collapse)

```python
arr = [[1, 2, 3],
       [4, 5, 6]]

np.sum(arr, axis=0)  → [5, 7, 9]    # Sum of each column
np.sum(arr, axis=1)  → [6, 15]      # Sum of each row
```

````python
# ============================================================
# TOPIC 10 EXAMPLE: Aggregation Functions (1D)
# ============================================================

import numpy as np

# Student test scores
scores = np.array([72, 85, 90, 60, 78, 95, 88, 55, 70, 82])
print("Scores:", scores)
print()

print("Sum      :", np.sum(scores))              # Total marks
print("Mean     :", np.mean(scores))             # Average marks
print("Minimum  :", np.min(scores))              # Lowest score
print("Maximum  :", np.max(scores))              # Highest score
print("Median   :", np.median(scores))           # Middle score (sorted)
print("Std Dev  :", np.round(np.std(scores), 2)) # How spread out the scores are
print("Variance :", np.round(np.var(scores), 2)) # Variance (std² )

print()
# Finding index of min and max
print("Index of Min score:", np.argmin(scores))  # Who got lowest?
print("Index of Max score:", np.argmax(scores))  # Who got highest?
````

````python
# ============================================================
# TOPIC 10 EXAMPLE: Aggregation on 2D Arrays (with axis)
# ============================================================

import numpy as np

# 3 students, 4 subjects
marks = np.array([
    [80, 90, 70, 85],   # Student 0: Maths, Science, English, Hindi
    [60, 75, 88, 92],   # Student 1
    [95, 85, 78, 70]    # Student 2
])

print("Marks Table (rows=students, cols=subjects):")
print(marks)
print()

# Total of everything
print("Overall Total  :", np.sum(marks))
print("Overall Mean   :", np.mean(marks))

print()

# axis=1 → sum across each ROW (total per student)
print("Total per Student (axis=1):", np.sum(marks, axis=1))

# axis=0 → sum down each COLUMN (total per subject)
print("Total per Subject (axis=0):", np.sum(marks, axis=0))

print()

# Average per student
print("Avg per Student (axis=1) :", np.round(np.mean(marks, axis=1), 2))

# Average per subject
print("Avg per Subject (axis=0) :", np.round(np.mean(marks, axis=0), 2))
````

## 📌 Topic 11: Boolean Masking / Filtering

### 🧑‍🏫 Theory

**Boolean masking** is a powerful way to **filter elements** from an array based on a **condition**.

Here's how it works in two steps:

**Step 1**: Apply a condition to the array → This gives you an array of `True`/`False` values (called a **boolean mask**).

```python
arr = np.array([10, 25, 5, 40, 15])
mask = arr > 15
# mask → [False, True, False, True, False]
```

**Step 2**: Use the mask to filter the original array — only `True` positions are kept:

```python
arr[mask]  → [25, 40]   # Only elements where mask is True
```

This can also be written in one line:

```python
arr[arr > 15]  → [25, 40]
```

---

### 🔗 Combining Conditions

Use `&` (AND) and `|` (OR) to combine multiple conditions.  
⚠️ Always wrap each condition in **parentheses `()`**!

```python
arr[(arr > 10) & (arr < 40)]   # Elements between 10 and 40
arr[(arr < 5) | (arr > 35)]    # Elements less than 5 OR greater than 35
```

````python
# ============================================================
# TOPIC 11 EXAMPLE: Boolean Masking / Filtering
# ============================================================

import numpy as np

scores = np.array([55, 72, 48, 90, 65, 38, 85, 77, 60, 92])
print("All Scores:", scores)
print()

# --- Step 1: Create a boolean mask ---
pass_mask = scores >= 60
print("Pass Mask (scores >= 60):", pass_mask)
# Output: [False True False True True False True True True True]

print()

# --- Step 2: Apply mask to filter ---
pass_scores = scores[pass_mask]
print("Passing Scores:", pass_scores)

# --- One-liner version ---
fail_scores = scores[scores < 60]
print("Failing Scores:", fail_scores)

print()

# --- Count of passing and failing students ---
print("Number of students who passed:", len(pass_scores))
print("Number of students who failed:", len(fail_scores))
````

````python
# ============================================================
# TOPIC 11 EXAMPLE: Combining Conditions
# ============================================================

import numpy as np

scores = np.array([55, 72, 48, 90, 65, 38, 85, 77, 60, 92])
print("Scores:", scores)
print()

# AND condition: Scores between 60 and 85 (Grade B students)
grade_b = scores[(scores >= 60) & (scores <= 85)]
print("Grade B (60–85):", grade_b)

# OR condition: Below 50 OR above 90 (Extremes)
extremes = scores[(scores < 50) | (scores > 90)]
print("Extremes (<50 or >90):", extremes)

# NOT condition: Scores that are NOT failing
not_failing = scores[~(scores < 60)]    # ~ means NOT
print("Not Failing (not <60):", not_failing)

print()

# Practical: Assign grade labels using boolean checks
print("Any score above 90?", np.any(scores > 90))    # True if at least one
print("All scores above 30?", np.all(scores > 30))   # True if all satisfy
````

## 📌 Topic 12: `np.where()`

### 🧑‍🏫 Theory

`np.where()` is like a **vectorized if-else statement**.  
It applies a condition element by element and returns one value if `True` and another if `False`.

### 🔤 Syntax

```python
np.where(condition, value_if_true, value_if_false)
```

### 🏫 Analogy

Think of it like checking every student's marks:

> *"If the score is 60 or more, write 'Pass'. Otherwise, write 'Fail'."*

With a regular Python loop, you'd need several lines of code.  
With `np.where()`, you do it in **one line**!

```python
np.where(scores >= 60, "Pass", "Fail")
```

---

### 📝 Two ways to use `np.where()`

**1. With condition + values** (if-else replacement):
```python
np.where(condition, true_value, false_value)
```

**2. With condition only** (returns indices where condition is True):
```python
np.where(condition)   # Returns a tuple of indices
```

````python
# ============================================================
# TOPIC 12 EXAMPLE: np.where()
# ============================================================

import numpy as np

scores = np.array([55, 72, 48, 90, 65, 38, 85, 77, 60, 92])
print("Scores:", scores)
print()

# --- Basic np.where(): Pass or Fail ---
result = np.where(scores >= 60, "Pass", "Fail")
print("Pass/Fail Result:", result)

print()

# --- Replace values using np.where() ---
# Apply +5 bonus to students who failed (scored < 60), leave others unchanged
adjusted = np.where(scores < 60, scores + 5, scores)
print("Original Scores :", scores)
print("Adjusted Scores  :", adjusted)  # Only failing students get +5

print()

# --- Assign Grade Labels ---
grades = np.where(scores >= 85, "A",
         np.where(scores >= 70, "B",
         np.where(scores >= 60, "C", "F")))

print("Scores:", scores)
print("Grades :", grades)
````

````python
# ============================================================
# TOPIC 12 EXAMPLE: np.where() to find indices
# ============================================================

import numpy as np

scores = np.array([55, 72, 48, 90, 65, 38, 85, 77, 60, 92])

# Find INDICES where scores > 80
indices = np.where(scores > 80)[0]
print("Indices where score > 80:", indices)      # Returns array of indices
print("Values at those indices  :", scores[indices]) # Get the actual values

print()

# Practical example: Find which students need remedial support
remedial_indices = np.where(scores < 50)[0]
print("Students needing remedial support (index):", remedial_indices)
print("Their scores:", scores[remedial_indices])
````

### ❓ Q8. Boolean Masking
Given temperatures: `temps = np.array([22, 35, 18, 40, 28, 15, 38, 25, 42, 30])`

- (a) Find all temperatures **above 30°C** (hot days).
- (b) Find all temperatures **between 20°C and 30°C** (comfortable days).
- (c) Count how many days were **below 20°C** (cold days).
- (d) Check if **any** day had temperature **above 40°C**.

````python
# Q8: Your answer here
import numpy as np

temps = np.array([22, 35, 18, 40, 28, 15, 38, 25, 42, 30])

# (a) Above 30°C
temp_above30 = temps[temps > 30]
print("Days above 30°C:", temp_above30)
# (b) Between 20 and 30
temp_between20_30 = temps[(temps > 20) & (temps < 30)]
print("Days between 20 and 30°C:", temp_between20_30)
# (c) Count of days below 20°C
count_below20 = np.sum(temps < 20)
print("Count of days below 20°C:", count_below20)

# (d) Any day above 40°C?
any_above40 = np.any(temps > 40)
print("Any day above 40°C?", any_above40)
````

### ❓ Q9. `np.where()` — Grade Assignment
Given scores: `scores = np.array([45, 62, 78, 91, 55, 83, 70, 38, 95, 66])`

Use `np.where()` to:
- (a) Label each score as `"Pass"` (≥ 60) or `"Fail"` (< 60).
- (b) Apply a **curved score**: Add 5 marks to anyone who scored **below 50**, leave others unchanged.
- (c) Assign grade **"Distinction"** if score ≥ 85, **"Merit"** if score ≥ 65, else **"Pass/Fail"** as appropriate.

````python
# Q9: Your answer here
import numpy as np

scores = np.array([45, 62, 78, 91, 55, 83, 70, 38, 95, 66])

# (a) Pass or Fail
mark_check = np.where(scores >= 60, "Pass", "Fail")
print("Pass/Fail Result:", mark_check)
# (b) Curved score (add 5 to scores < 50)
mark_add_5 = np.where(scores < 60, scores + 5, scores)
print("Curved Scores:", mark_add_5)
# (c) Grade assignment
assign_grades = np.where(scores >= 85, "Distinction",
                 np.where(scores >= 65, "Merit",
                 np.where(scores >= 70, "Pass", "Fail")))
print("Grade Assignment:", assign_grades)
````

### ❓ Q10. Mini Challenge — Put It All Together! 🏆

You are a teacher with the following data for **6 students** across **4 subjects**:  
*(Subjects: Maths, Science, English, Computer)*

```
Student 0: [82, 76, 90, 88]
Student 1: [55, 60, 72, 50]
Student 2: [91, 85, 78, 95]

```

Do the following:
- (a) Print the shape of the array.
- (b) Find the **total marks** for each student.
- (c) Find the **average marks** per subject.
- (d) Identify students whose **total marks are below 250** (struggling students).
- (e) Use `np.where()` to label each student's **total** as `"Good"` if ≥ 300, else `"Needs Improvement"`.

````python
# Q10: Your answer here
import numpy as np

marks = np.array([
    [82, 76, 90, 88],   # Student 0
    [55, 60, 72, 50],   # Student 1
    [91, 85, 78, 95],   # Student 2
    [40, 52, 61, 45],   # Student 3
    [70, 68, 80, 74],   # Student 4
    [88, 92, 85, 79]    # Student 5
])

# (a) Shape
print(marks.shape)
# (b) Total per student
sum_of_marks = np.sum(marks, axis=1)
print("Total Marks per Student:", sum_of_marks)
# (c) Average per subject
print(np.mean(marks, axis=0))
# (d) Struggling students (total < 250)
print("Struggling Students (Total < 250):", np.where(sum_of_marks < 250)[0])
# (e) Label each student
label_student = np.where(sum_of_marks >= 300, "Good", "Needs Improvement")
print("Student Labels:", label_student)
````
