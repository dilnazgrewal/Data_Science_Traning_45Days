```python
df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")
df["Discount"] = pd.to_numeric(df["Discount"], errors="coerce")
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
df["Delivery_Days"] = pd.to_numeric(df["Delivery_Days"], errors="coerce")
df.dtypes
```

```python
df_clean = df.copy()
```

```python
df_clean.isnull().sum()
```

```python
df_clean.dtypes
```

```python
df_clean["Customer_Name"] = df_clean["Customer_Name"].fillna("Unknown")
df_clean["Customer_Name"].isnull().sum()
```

```python
df_clean['Category'] = df_clean['Category'].fillna(df_clean['Category'].mode()[0])
df_clean["Category"].isnull().sum()
```

```python
df_clean['City'] = df_clean['City'].fillna(df_clean['City'].mode()[0])
df_clean["City"].isnull().sum()
```

```python
df_clean.isnull().sum()
```

```python
median_days = round(df_clean['Delivery_Days'].median(),1)
df_clean['Delivery_Days'] = df_clean['Delivery_Days'].fillna(median_days)
```

```python
median_val = df_clean['Rating'].median()
df_clean['Rating'] = df_clean['Rating'].fillna(median_val)
```

```python
# df_clean = df_clean.dropna(subset="Sales")
median_sale = df_clean['Sales'].median()
print(median_sale)
df_clean['Sales'] = df_clean['Sales'].fillna(median_sale)
```

```python
df_clean["Discount"] = df_clean["Discount"].fillna(df_clean["Discount"].median())
```

```python
df_clean['Quantity'] = df_clean['Quantity'].fillna(df_clean["Quantity"].median())
```

```python
print(df_clean.isnull().sum())
```

```python
df_clean = df_clean.drop_duplicates()
print("Duplicate values before:", len(df[df.duplicated()]))
print("Duplicate values after treatment:", len(df_clean[df_clean.duplicated()]))
```

```python
df_clean.info()
```

```python
df_clean['Category'] = df_clean['Category'].str.strip().str.title()
df_clean['Customer_Name'] = df_clean['Customer_Name'].str.strip().str.title()
df_clean['City'] = df_clean['City'].str.strip().str.title()
df_clean['Product_Name'] = df_clean['Product_Name'].str.strip().str.title()
df_clean['Payment_Mode'] = df_clean['Payment_Mode'].str.strip().str.title()
df_clean['Returned'] = df_clean['Returned'].str.strip().str.title()
```

```python
for cols in ['Category', 'City', 'Payment_Mode', 'Product_Name', 'Customer_Name', 'Returned']:
    print(f"Unique values in '{cols}': {df_clean[cols].unique().tolist()}\n\n")
```

```python
df_clean['Order_Date'].head(30)
```

```python
df_clean['Order_Date'] = pd.to_datetime(df_clean['Order_Date'], format = "mixed", errors="coerce")

print("Invalid dates:", df_clean["Order_Date"].isnull().sum())
```

```python
df_clean.info()
```

```python
df_clean.loc[df_clean['Quantity'] < 0, 'Quantity'] = np.nan

df_clean.loc[(df_clean['Discount'] < 0) | (df_clean['Discount'] > 1), 'Discount'] = np.nan

df_clean.loc[(df_clean['Rating'] < 0) | (df_clean['Rating'] > 5), 'Rating'] = np.nan

columns_to_fill = ["Quantity", "Discount", "Rating"]

for column in columns_to_fill:
    df_clean[column] = df_clean[column].fillna(df_clean[column].median())
print("INVALID VALUES HANDLED")
```

```python
cols = ['Discount', 'Quantity', 'Sales', 'Rating']
for col in cols:
    
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)]
    print(f"\n→ Found {len(outliers)} outliers in {col}")
```

```python
outlier_columns = ["Discount", "Rating", "Sales"]

for col in outlier_columns:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_clean[col] = df_clean[col].clip(
        lower=lower_bound,
        upper=upper_bound
    )

    outliers_after_capping = df_clean[
        (df_clean[col] < lower_bound) |
        (df_clean[col] > upper_bound)
    ]

    print(f"{col}: Outliers after capping = {len(outliers_after_capping)}")
```

```python
df_clean['Total_Amount'] = df_clean['Sales'] * df_clean['Quantity']

df_clean['Discount_Amount'] = df_clean['Total_Amount'] * df_clean['Discount']

df_clean['Final_Amount'] = df_clean['Total_Amount'] - df_clean['Discount_Amount']

df_clean['Order_Year'] = df_clean['Order_Date'].dt.year

df_clean['Order_Month'] = df_clean['Order_Date'].dt.month_name()

df_clean['Order_Day'] = df_clean['Order_Date'].dt.day_name()
```

```python
# What is the total sales amount?
print(f"Total Sales: Rs.{round(df_clean['Sales'].sum(),3)}")
# Which category has the highest total sales?
print(f"Highest sales category: {df_clean.groupby('Category')['Sales'].sum().idxmax()}")
# Which region has the highest sales?
print(f"Highest sales region: {df_clean.groupby('City')['Sales'].sum().idxmax()}")
# Which product has the highest quantity sold?
print(f"Highest Quantity sold Product: {df_clean.groupby('Product_Name')['Quantity'].sum().idxmax()}")
# What is the average rating?
print(f"Average Rating: {df_clean['Rating'].mean()}")
# Which payment method is used most?
print(f"Most used payment method: {df_clean['Payment_Mode'].value_counts().idxmax()}")
# Which month has the highest sales?
print(f"Month with highest sales: {df_clean.groupby('Order_Month')['Sales'].sum().idxmax()}")
# Find the top 5 highest sales records.
print(f"Highest Sales Record:")
display(df_clean.sort_values('Sales', ascending= False).head())
# Find the bottom 5 sales records.
print(f"Bottom Sales Record:")
display(df_clean.sort_values('Sales', ascending= True).head())
# Count the number of orders in each category.
print(f"Number of Orders in each category: {df_clean['Category'].value_counts()}")
```

```python
# Total sales by category
print("Total sales by category:\n", df_clean.groupby('Category')['Sales'].sum())
# Total sales by region
print("\nTotal sales by region:\n", df_clean.groupby('City')['Sales'].sum())
# Average rating by category
print("\nAverage Rating by category:\n", df_clean.groupby('Category')['Rating'].mean())
# Total quantity sold by product
print("\nTotal quantity sold by product:\n", df_clean.groupby('Product_Name')['Quantity'].sum())
# Average final amount by payment method
print("\nAverage final amount by payment method:\n", df_clean.groupby('Payment_Mode')['Final_Amount'].mean())
```
