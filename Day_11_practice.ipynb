## 📌 Key Takeaways

Here's a summary of everything we learned today:

---

### 1. Loading Data
- Use `pd.read_excel()` for Excel files and `pd.read_csv()` for CSV files.
- The result is stored in a **DataFrame** (a table).

### 2. Viewing Data
| Method | What It Does |
| :--- | :--- |
| `.head(n)` | First `n` rows (default 5) |
| `.tail(n)` | Last `n` rows (default 5) |
| `.sample(n)` | `n` random rows |

### 3. Understanding Structure
| Method | What It Returns |
| :--- | :--- |
| `.shape` | (rows, columns) — **no parentheses!** |
| `.columns` | List of all column names |
| `.dtypes` | Data type of each column |
| `.info()` | Complete summary with types and null counts |

### 4. Statistical Summary
| Method | What It Returns |
| :--- | :--- |
| `.describe()` | Stats for numeric columns |
| `.describe(include='all')` | Stats for ALL columns |
| `.sum()` | Total of a column |
| `.mean()` | Average of a column |
| `.min()` / `.max()` | Smallest / Largest value |
| `.median()` | Middle value |

### 5. Exploring Values
| Method | What It Returns |
| :--- | :--- |
| `.unique()` | All distinct values |
| `.nunique()` | Count of distinct values |
| `.value_counts()` | Frequency of each value |
| `.isnull().sum()` | Count of missing values |

### 6. Sorting
- `.sort_values('column')` — ascending (default)
- `.sort_values('column', ascending=False)` — descending

### 7. Selecting Columns
- Single column: `df['Column']` → returns a **Series**
- Multiple columns: `df[['Col1', 'Col2']]` → returns a **DataFrame**
