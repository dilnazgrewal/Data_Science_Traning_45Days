# Module 2: Data Science Lifecycle

Every successful Data Science project follows a structured, step-by-step process. This is known as the **Data Science Lifecycle**.

```text
    Business Problem
           │
           ▼
    Data Collection
           │
           ▼
     Data Cleaning
           │
           ▼
     Data Analysis
           │
           ▼
    Model Building
           │
           ▼
      Evaluation
           │
           ▼
       Deployment
           │
           ▼
       Monitoring
```

Let's break down each step using a real-world example: **A Telecom company wants to reduce customer churn (customers leaving the service).**

---

### Step 1: Business Problem 🎯
First, we must define the problem and ask the right questions.
* **Example:** The telecom company wants to identify customers likely to leave.
* **Core Question:** *"Which customers are likely to churn (cancel their subscription)?"*

### Step 2: Data Collection 📥
Next, we gather relevant data from various sources.
* **Data Sources:**
  * Databases (SQL databases)
  * Excel files or CSV spreadsheets
  * APIs (web services)
  * Websites (web scraping)
  * Mobile applications (app logs)

### Step 3: Data Cleaning 🧹
Real-world data is messy, incomplete, and contains errors. We must clean it.
* **Common Problems:**
  * **Missing Values:** Age is unknown (`?` or `NaN`), or Salary is `NULL`.
  * **Duplicate Records:** The same customer is listed twice in the database.
  * **Incorrect Entries:** Age is listed as `-5` or a negative salary.

### Step 4: Data Analysis (EDA) 📊
We explore the clean data to find initial trends, summaries, and insights.
* **Questions we might ask:**
  * What is the average age of our customers?
  * Which data plans sell the most?
  * Which city generates the highest revenue?
  * Do customers with higher bills leave more often?

### Step 5: Model Building 🤖
We use historical data to build mathematical or machine learning models that can make predictions.
* **Example:** Build a predictive model that takes customer usage details and predicts whether they will churn (Yes or No).

### Step 6: Evaluation 📈
We test the model on new data to see how accurate it is before deployment.
* **Example:** Checking if the model is correct. *"Our churn model is 92% accurate!"*

### Step 7: Deployment 🚀
We integrate the model into a live application where business users can use it.
* **Example:** Deploying the model to the customer service system so agents see an alert when a churning customer calls.

### Step 8: Monitoring 🔍
We continuously monitor the model's performance over time. Because customer behavior and business environments change, the model must be periodically updated.

# Module 4: Python Fundamentals - Variables

## What is a Variable?
A **variable** is a container used to store data in a computer's memory. Think of a variable as a **labeled storage box**. You put data inside the box and give the box a name so you can find it later.

Here is how we declare variables in Python:
```python

age = 25              # Stores a whole number (Integer)
salary = 50000        # Stores an integer or decimal number
```

---

## Variable Naming Rules ⚠️
To make sure Python understands your variables, follow these naming rules:

### ✅ Valid Names:
* `student_name = "Rahul"` (Can use underscores)
* `age1 = 25` (Can contain numbers, but not start with them)
* `salary_amount = 50000` (Descriptive names)

### ❌ Invalid Names:
* `1age = 25` (Cannot start with a number)
* `student-name = "Rahul"` (Cannot use hyphens `-`)
* `student name = "Rahul"` (Cannot contain spaces)

### 💡 Best Practices:
* **Use meaningful and descriptive names:** Write `employee_salary` instead of just `s`.
* **Use lowercase with underscores (snake_case):** This is the standard style for Python variables.


````python
# Declaring variables
student_name = "Rahul"
Age = 25
salary_amount = 50000

# Printing the variables to see their values
print(student_name,Age,salary_amount)
print("Student Name:", student_name)
print("Age:", age)
print("Salary Amount:", salary_amount)
````

# Module 6: Python Data Types

## Overview
A **data type** defines the kind of value stored in a variable. In Data Science, understanding data types helps us process clean datasets.

We will cover six core data types:
1. **Integer (`int`):** Whole numbers (e.g., 25, -100).
2. **Float (`float`):** Decimal numbers (e.g., 45000.75, 3.14).
3. **String (`str`):** Text data (e.g., "Rahul").
4. **List (`list`):** Ordered collections of items (e.g., `["Apple", "Mango"]`).
5. **Tuple (`tuple`):** Ordered, fixed collections of items (e.g., `(1, 2, 3)`).
6. **Set (`set`):** Unordered collections of unique items (e.g., `{"Apple", "Mango"}`).

---

## Mutable vs Immutable 🔄
* **Immutable (Cannot change):** Once created, their values cannot be altered. If you modify them, Python creates a new object in memory.
  * *Examples:* `int`, `float`, `str`, `tuple`
* **Mutable (Can change):** You can add, remove, or modify items after creation without creating a new object in memory.
  * *Examples:* `list`, `dict`, `set`

### Mutability Summary Table
| Data Type | Keyword | Mutable / Immutable |
| :--- | :--- | :--- |
| **Integer** | `int` | **Immutable** |
| **Float** | `float` | **Immutable** |
| **String** | `str` | **Immutable** |
| **List** | `list` | **Mutable** |
| **Tuple** | `tuple` | **Immutable** |
| **Set** | `set` | **Mutable** |
| **Dictionary** | `dict` | **Mutable** |

---

## 1. Integer (`int`) & Float (`float`)
* **Integer (`int`):** Stores whole numbers. Immutable.
* **Float (`float`):** Stores decimal numbers. Immutable.

### Common Functions:
* `type()`: Check data type.
* `abs()`: Absolute value.

* `pow(base, exp)`: Power calculation.
* `round(val, decimals)`: Rounding decimals.

````python
age = 25
salary = 45000.75

print("Type of age:", type(age))
print("Absolute value of -25:", abs(-25))
print("Age squared:", pow(age, 3))

print("\n--- Float Operations ---")
print("Type of salary:", type(salary))
print("Round salary (nearest integer):", round(salary))
print("Round salary to 1 decimal place:", round(salary, 1))
````

## 2. String (`str`)
Strings store text. They are written inside single quotes `'...'` or double quotes `"..."`.

### Indexing & Slicing 
Since strings are ordered sequences of characters, we can extract specific parts:
* **Indexing:** Accessing a single character by its position (starts at `0` for the first character, and `-1` for the last character).
* **Slicing:** Accessing a range of characters using `[start:end]` (where the `end` position is excluded).

### Common String Functions:
* `len(text)`: Returns the length (number of characters).
* `text.upper()`: Converts text to UPPERCASE.
* `text.lower()`: Converts text to lowercase.
* `text.replace(old, new)`: Replaces a word/character with another.
* `text.startswith(prefix)`: Checks if the string starts with a specific prefix (returns `True` or `False`).


````python
name = "Rahul"

print("--- Indexing ---")
print("First letter name[0]:", name[0])
print("Last letter name[-1]:", name[-1])

print("\n--- Slicing ---")
print("First 3 letters name[0:3]:", name[0:3])

print("\n--- Common Functions ---")
print("Length of name:", len(name))
print("Uppercase name:", name.upper())
print("Lowercase name:", name.lower())
print("Replace 'Rahul' with 'Amit':", name.replace("Rahul", "Amit"))
print("Starts with 'Ra'?:", name.startswith("Ra"))
````

## 3. List (`list`)
A **List** stores multiple items in a single variable. Lists are ordered, allow duplicates, and are mutable. They are written with square brackets `[...]`.

### Slicing Lists:
Works just like string slicing! `fruits[0:2]` gets items at index 0 and 1.

### Common List Functions:
* `list.append(item)`: Adds an item to the end.
* `list.insert(index, item)`: Adds an item at a specific position.
* `list.remove(item)`: Removes the first occurrence of an item.
* `list.sort()`: Sorts items alphabetically or numerically.
* `len(list)`: Checks the number of items.


````python
fruits = ["Apple", "Mango", "Banana"]

print("--- Original List ---")
print(fruits)

print("\n--- Indexing & Slicing ---")
print("First fruit fruits[0]:", fruits[0])
print("First two fruits fruits[0:2]:", fruits[0:2])

print("\n--- List Modifications (Mutable!) ---")
fruits.append("Orange")
print("After append('Orange'):", fruits)

fruits.insert(1, "Kiwi")
print("After insert(1, 'Kiwi'):", fruits)

fruits.remove("Banana")
print("After remove('Banana'):", fruits)

fruits.sort()
print("After sort():", fruits)
print("Length of fruits list:", len(fruits))
````

## 4. Dictionary (`dict`)
A **Dictionary** stores data as key-value pairs (like a real word dictionary: the key is the word, and the value is the definition). Dictionaries are written inside curly braces `{key: value}`. Keys must be unique.

### Common Dictionary Functions:
* `dict.get(key)`: Safely retrieves the value of a key.
* `dict.keys()`: Returns a list of all keys.
* `dict.values()`: Returns a list of all values.
* **Add new key:** `dict[new_key] = value`


````python
student = {
    "name": "Rahul",
    "age": 22
}

print("--- Accessing dictionary values ---")
print("Using get('name'):", student.get("name"))
print("Using square brackets student['age']:", student["age"])

print("\n--- Keys and Values ---")
print("Keys:", list(student.keys()))
print("Values:", list(student.values()))

print("\n--- Adding a new key-value pair (Mutable!) ---")
student["city"] = "Delhi"
print("Updated student dictionary:", student)
````

## 5. Tuple (`tuple`)
A **Tuple** stores multiple items in a single variable, just like a list. However, tuples are **ordered** and **immutable**. They are written with parentheses `(...)`.

### Key Tuple Features:
* Ordered collection: Items keep their position.
* Immutable: You cannot add, remove, or change items after creation.
* Allows duplicates.

### Common Tuple Operations:
* `tuple[index]`: Access an item by position.
* `len(tuple)`: Get the number of items.
* `tuple.count(value)`: Count how many times a value appears.
* `tuple.index(value)`: Find the position of a value.

````python
student_scores = (85, 90, 78, 90)

print("--- Tuple Example ---")
print("student_scores:", student_scores)
print("First score student_scores[0]:", student_scores[0])
print("Length of tuple:", len(student_scores))
print("Count of 90 in tuple:", student_scores.count(90))
print("Index of 78 in tuple:", student_scores.index(78))
````

## 6. Set (`set`)
A **Set** stores multiple items in a single variable with two important rules: items are **unordered** and **unique**. Sets are written with curly braces `{...}`.

### Key Set Features:
* Unordered collection: Items have no fixed position.
* Unique items only: Duplicates are automatically removed.
* Mutable: You can add or remove items.

### Common Set Operations:
* `set.add(item)`: Add an item.
* `set.remove(item)`: Remove an item.
* `set.union(other_set)`: Combine two sets.
* `set.intersection(other_set)`: Find common items.
* `set.difference(other_set)`: Find items in one set but not the other.

````python
fruits_set = {"Apple", "Mango", "Banana", "Apple"}
print("--- Set Example ---")
print("fruits_set:", fruits_set)

fruits_set.add("Orange")
print("After add('Orange'):", fruits_set)

fruits_set.remove("Mango")
print("After remove('Mango'):", fruits_set)

other_fruits = {"Banana", "Grapes", "Kiwi"}
print("Union:", fruits_set.union(other_fruits))
print("Intersection:", fruits_set.intersection(other_fruits))
print("Difference:", fruits_set.difference(other_fruits))
````

# Module 7: Operators

Operators are symbols used to perform calculations, compare values, and check conditions.

---

## 1. Arithmetic Operators
Used for standard mathematical calculations.

| Operator | Meaning | Example (a = 10, b = 3) | Result |
| :---: | :--- | :--- | :---: |
| `+` | Addition | `a + b` | `13` |
| `-` | Subtraction | `a - b` | `7` |
| `*` | Multiplication | `a * b` | `30` |
| `/` | Division | `a / b` | `3.3333...` |
| `%` | Modulus (Remainder) | `a % b` | `1` |
| `//` | Floor Division (Rounds down) | `a // b` | `3` |
| `**` | Power (Exponent) | `a ** b` | `1000` |


````python
a = 10
b = 3

print("Addition a + b:", a + b)
print("Subtraction a - b:", a - b)
print("Multiplication a * b:", a * b)
print("Division a / b:", a / b)
print("Modulus (Remainder) a % b:", a % b)
print("Floor Division a // b:", a // b)
print("Power a ** b:", a ** b)
````

## 2. Membership Operators
Used to check if an item is inside a list or collection.
* `in`: Returns `True` if found.
* `not in`: Returns `True` if NOT found.


````python
fruits = ["Apple", "Mango", "Banana"]

print("Is 'Mango' in fruits?:", "Mango" in fruits)
print("Is 'Orange' not in fruits?:", "Orange" not in fruits)
````

## 3. Identity Operators (`is` vs `==`)
Identity operators check if two variables reference the exact same block of memory.

### 💡 Crucial Difference:
* **Equality (`==`):** Checks if the **values** inside the variables are equal.
* **Identity (`is`):** Checks if the variables point to the **exact same memory location**.


````python
# Scenario 1: Variables pointing to the same memory reference
list1 = [10, 20, 30]
list2 = list1  # list2 now references the exact same list1

print("--- Scenario 1 ---")
print("Equality list1 == list2:", list1 == list2)   # True (values are the same)
print("Identity list1 is list2:", list1 is list2)   # True (same memory location)

# Scenario 2: Two separate lists with the same values
list3 = [10, 20, 30]
list4 = [10, 20, 30]  # Created separately

print("\n--- Scenario 2 ---")
print("Equality list3 == list4:", list3 == list4)   # True (values are the same)
print("Identity list3 is list4:", list3 is list4)   # False (different locations in memory)
print("Identity list3 is not list4:", list3 is not list4) # True
````
