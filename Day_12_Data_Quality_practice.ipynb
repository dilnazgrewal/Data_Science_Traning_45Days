```python
import pandas as pd
```

```python
df = pd.read_excel("DataSet_used\\sales_data_messy.xlsx")
```

```python
df.head()
```

```python
print("\nMISSING VALUES — COUNT & PERCENTAGE")
missing_count = df.isnull().sum()  
missing_percent = (missing_count / len(df)) * 100  

missing_summary = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing %': missing_percent.round(2)})
print(missing_summary)
```

```python
print("Duplicate rows:",df.duplicated().sum())
```

```python
print("Price Analysis")
print("="*30)
print("Minimum Price: Rs.", df['Price'].min())
print("Maximum Price: Rs.", df['Price'].max())
print("Average Price: Rs.", round(df['Price'].mean(),2))
print("Negative Prices: Rs.", (df['Price'] < 0).sum())
print("Zero Prices:", (df['Price'] == 0).min().sum())
print("="*30)
print("\n")
print("Quantity Analysis")
print("="*30)
print("Minimum Quantity:", df['Quantity'].min())
print("Maximum Quantity:", df['Quantity'].max())
print("Average Quantity:", round(df['Quantity'].mean(),2))
print("Negative Quantities:", (df['Quantity'] < 0).sum())
print("Zero Quantities:", (df['Quantity'] == 0).min().sum())
print("="*30)
print("\n")
print("Revenue Analysis")
print("="*30)
print("Minimum Revenue:", df['Revenue'].min())
print("Maximum Revenue:", df['Revenue'].max())
print("Average Revenue:", round(df['Revenue'].mean(),2))
print("Negative Revenue:", (df['Revenue'] < 0).sum())
print("Zero Revenue:", (df['Revenue'] == 0).min().sum())
print("="*30)
```

```python
print("Unique Products:",df['Product'].unique().tolist())
print("Unique Categories:",df['Category'].unique().tolist())
```

```python
df["Date"] = pd.to_datetime(df["Date"])
print(df["Date"].dtype)
print("Time span:", df["Date"].max().date() - df["Date"].min().date())
```

```python
df["Expected_Revenue"] = df["Price"] * df["Quantity"]
```

```python
mismatch = df[df["Expected_Revenue"] != df['Revenue']]
print("TOTAL MISMATCHES:",len(mismatch))
print()
display(mismatch[["Date", "Category", "Product", "Revenue", "Expected_Revenue"]])
```

```python
print(f"Current Column (Before): {df.columns.tolist()}")
df.drop(columns ='Expected_Revenue', inplace = True)
print(f"\nCurrent Column (After): {df.columns.tolist()}")
```
