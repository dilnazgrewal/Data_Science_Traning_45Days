```python
x = [10,20]
y = [10,20]
```

```python
if x == y:
    print("True")
else:
    print("False")
```

```python
if x is y: 
    print("Both are identical")
else:
    print("Both are not identical")
```

```python
z = x
if z is x: 
    print("Both are identical")
else:
    print("Both are not identical")
```

```python
allowed_domains = ["gmail.com", "yahoo.com", "outlook.com"]

string = "john.doe@gmail.com"
stripped = string.split("@")[1]
if stripped in allowed_domains:
    print("User has an allowed domain.")
else:
    print("User does not have an allowed domain.")
```

```python
year = int(input("Enter Year:"))
if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
    print("Leap Year")
else:
    print("Not a Leap Year") 
```
