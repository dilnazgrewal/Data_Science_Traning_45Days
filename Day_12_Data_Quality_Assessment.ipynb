## 📖 Concept Explanation

### What is Data Quality?

**Data Quality** refers to how accurate, complete, and reliable your data is. Before doing any analysis, we need to answer:

| Question | Why It Matters |
|----------|----------------|
| Are there **missing values**? | Missing data can give wrong averages, wrong counts, and misleading charts |
| Are there **duplicate rows**? | Duplicates inflate numbers — e.g., counting the same sale twice |
| Are the **data types correct**? | A date stored as text cannot be used for time-based analysis |
| Are **numerical values sensible**? | Negative prices or quantities of 0 indicate errors |
| Are **categorical values consistent**? | "Electronics" vs "electronics" vs "ELECTRONICS" are treated as different |
| Is the **data consistent**? | If Revenue should equal Price × Quantity, we must verify it |

### The Data Quality Assessment Workflow

```
Step 1: Load the Data
    ↓

Step 6: Inspect Categorical Columns
    ↓
Step 7: Inspect Date Columns
    ↓
Step 8: Verify Data Consistency
    ↓

```

### 🔍 1. Check Missing Values

**What are Missing Values?**

Missing values are blank or empty cells in the data. In Python, they appear as `NaN` (Not a Number) or `None`.

**Why are they a problem?**
- They can give wrong averages (e.g., average salary ignoring some employees)
- Many functions crash or give errors with missing data
- Charts and visualizations become misleading

````python
# ---- CHECK MISSING VALUES ----

# .isnull() marks each cell as True (missing) or False (not missing)
# .sum() counts the number of True values per column

missing_values = df.isnull().sum()
print(missing_values)
````

````python
# Let's also calculate the PERCENTAGE of missing values
# This is more useful when the dataset has thousands of rows
# Formula: (missing count / total rows) * 100
print("\nMISSING VALUES — COUNT & PERCENTAGE")
missing_count = df.isnull().sum()             # Count of missing per column
missing_percent = (missing_count / len(df)) * 100  # Percentage of missing

# Create a summary table
missing_summary = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing %': missing_percent.round(2)  # Round to 2 decimal places
})

print(missing_summary)
````

### 🔍 2. Check Duplicate Records

**What are Duplicates?**

A duplicate row is an **exact copy** of another row. For example, the same transaction recorded twice.

**Why are they a problem?**
- Revenue gets counted twice → wrong total sales
- Product counts get inflated
- Reports become inaccurate

````python
# ---- CHECK DUPLICATE RECORDS ----

# .duplicated() marks each row as True (duplicate) or False (not duplicate)
# .sum() counts the total number of duplicate rows

duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
````

**Expected Output:**
```
==================================================
DUPLICATE RECORDS CHECK
==================================================
Number of duplicate rows: 0
```

````python
# If duplicates existed, we can VIEW them using:
# df[df.duplicated()] → This filters and shows only the duplicate rows

# Let's check anyway (good practice)
duplicates = df[df.duplicated()]
print(f"Duplicate rows found: {len(duplicates)}")

display(duplicates)
````

````python
# We can also check duplicates based on a SPECIFIC column
# For example: Are there duplicate Transaction IDs?
# Each transaction should have a UNIQUE ID

txn_duplicates = df.duplicated(subset=['Transaction_ID']).sum()
print(f"Duplicate Transaction IDs: {txn_duplicates}")

if txn_duplicates == 0:
    print("✅ All Transaction IDs are unique!")
else:
    print("⚠️ Warning: Some Transaction IDs are repeated!")
df.drop_duplicates(inplace=True)
````

**Expected Output:**
```

✅ All Transaction IDs are unique!
```

> ✅ **Result:** No duplicate records found. Each transaction is unique.

**Questions to Students:**

### 🔍 5. Inspect Date Columns

**Why check Date columns?**

Dates are often loaded as **text (object type)** instead of **datetime**. If dates are stored as text:
- You CANNOT sort by date properly
- You CANNOT extract month, year, day of week
- You CANNOT do time-based analysis (e.g., monthly sales trends)

````python
# ---- INSPECT DATE COLUMN ----

# First, let's check the current data type of the Date column
print("=" * 50)
print("DATE COLUMN INSPECTION")
print("=" * 50)
print(f"Data type of 'Date' column: {df['Date'].dtype}")

# Show the first few date values
print(f"\nSample date values:")
print(df['Date'].head())
````

### 🔍 6. Verify Data Consistency

**What is Data Consistency?**

Data consistency means values across columns should **logically agree with each other**.

In our dataset: **Revenue should always equal Price × Quantity**

If `Revenue ≠ Price × Quantity`, it means there's a calculation error in the data.

````python
# ---- VERIFY DATA CONSISTENCY ----

# Check if Revenue = Price × Quantity for every row
# Create a new column with the expected revenue
df['Expected_Revenue'] = df['Price'] * df['Quantity']

# Compare actual Revenue with Expected Revenue
# If they don't match, there's a data quality issue
mismatches = df[df['Revenue'] != df['Expected_Revenue']]

print("=" * 50)
print("DATA CONSISTENCY CHECK")
print("=" * 50)
print(f"Mismatches found  : {len(mismatches)}")

if len(mismatches) == 0:
    print("\n✅ All Revenue values are consistent (Revenue = Price × Quantity)")
else:
    print("\n⚠️ Inconsistent rows found:")
    print(mismatches[[ 'Price', 'Quantity', 'Revenue', 'Expected_Revenue']])
````

````python
# Clean up: Remove the helper column we created
df.drop(columns=['Expected_Revenue'],inplace=True)

print("✅ Helper column removed. DataFrame is clean.")
print(f"Current columns: {df.columns.tolist()}")
````

> ✅ **Result:** Revenue = Price × Quantity for all 100 rows. The data is consistent!

## ⚠️ Common Mistakes

Here are mistakes beginners commonly make when assessing data quality.

---

### ❌ Mistake 1: Using `.isnull()` without `.sum()`

````python
# ❌ WRONG — This shows True/False for EVERY cell (huge output!)
# df.isnull()          # Shows a table of True/False — not useful

# ✅ CORRECT — This gives the COUNT of missing values per column
print("✅ Correct way to check missing values:")
print(df.isnull().sum())
````

---

### ❌ Mistake 2: Forgetting to convert Date column

````python
# ❌ WRONG — Trying to extract month from a text column
# This would give an error if Date is still 'object' type:
# df['Date'].dt.month   → AttributeError!

# ✅ CORRECT — First convert, then extract
# df['Date'] = pd.to_datetime(df['Date'])  # Convert first (already done above)

# Now we can safely extract month
print("✅ Extracting month after conversion:")
print(df['Date'].dt.month.head())
````

---

### ❌ Mistake 3: Confusing `.duplicated()` and `.drop_duplicates()`

````python
# .duplicated() → Returns True/False for each row (used to CHECK)
# .drop_duplicates() → Returns a new DataFrame WITHOUT duplicate rows (used to REMOVE)

# ❌ WRONG — Using drop_duplicates() just to count duplicates
# len(df) - len(df.drop_duplicates())   # Works but is roundabout

# ✅ CORRECT — Use .duplicated().sum() to count
print(f"✅ Number of duplicates: {df.duplicated().sum()}")
print("\nUse .duplicated().sum() to COUNT")
print("Use .drop_duplicates() to REMOVE")
````

---

### ❌ Mistake 4: Not checking data types before analysis

````python
# ❌ WRONG — Jumping straight into analysis without checking types
# df['Price'].mean()  → This works, but what if Price was stored as text?

# ✅ CORRECT — Always check dtypes first
print("✅ Always check data types before analysis:")
print(df.dtypes)
print("\n💡 TIP: Run df.dtypes or df.info() BEFORE doing any calculations!")
````

---

### ❌ Mistake 5: Using `=` instead of `==` for comparison

````python
# ❌ WRONG — Using single = (this ASSIGNS a value, doesn't compare!)
# df['Price'] = 0     → This would SET all prices to 0!

# ✅ CORRECT — Using double == (this COMPARES values)
zero_prices = (df['Price'] == 0).sum()   # This COUNTS rows where Price is 0
print(f"✅ Rows where Price equals 0: {zero_prices}")

print("\n💡 REMEMBER:")
print("   =  means ASSIGN (set a value)")
print("   == means COMPARE (check if equal)")
````
