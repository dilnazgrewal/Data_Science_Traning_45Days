```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Superstore_Messy.csv')

print("📊 STARTING THE CLEANING PIPELINE")
print("=" * 60)
print(f"   Rows:           {len(df)}")
print(f"   Missing values: {df.isnull().sum().sum()}")
print(f"   Duplicates:     {df.duplicated().sum()}")
print(f"   Date type:      {df['Order Date'].dtype}")
print("=" * 60)
```

```python
clean_df = df.copy()
clean_df.isnull().sum()
```

```python
print("-------- Handling missing values in Customer Names --------")
print("Missing values in Customer Name BEFORE handling:", clean_df["Customer Name"].isnull().sum())
clean_df["Customer Name"] = clean_df["Customer Name"].fillna("Unknown")
print("Missing values in Customer Name AFTER handling:", clean_df["Customer Name"].isnull().sum())

print("\n-------- Handling missing values in Postal Code --------")
print("Missing values in Postal Code BEFORE handling:", clean_df["Postal Code"].isnull().sum())
clean_df = clean_df.dropna(subset = ["Postal Code"])
print("Missing values in Postal Code AFTER handling:", clean_df["Postal Code"].isnull().sum())

print("\n-------- Handling missing values in Sales and Profit --------")
print("Missing values in Sales BEFORE handling:", clean_df["Sales"].isnull().sum())
print("Missing values in Profit BEFORE handling:", clean_df["Profit"].isnull().sum())
for col in ["Sales", "Profit"]:
    clean_df[col] = clean_df[col].fillna(clean_df[col].median())
print("\nMissing values in Sales AFTER handling:", clean_df["Sales"].isnull().sum())
print("Missing values in Profit AFTER handling:", clean_df["Profit"].isnull().sum())

```

```python
print("AFTER DEALING WITH MISSING VALUES:\n", clean_df.isnull().sum())
```

```python
print("Rows before dropping duplicates:", len(clean_df))
clean_df = clean_df.drop_duplicates()
print("Rows after deleting duplicates:", len(clean_df))
```

```python
print("INITIAL DATA TYPE OF DATES IN DATASET")
print(clean_df[["Order Date", "Ship Date"]].dtypes)

print("\nCONVERTING TO DATETIME")
clean_df["Order Date"] = pd.to_datetime(clean_df["Order Date"], format = "mixed")
clean_df["Ship Date"] = pd.to_datetime(clean_df["Ship Date"], format = "mixed")
print("\nAFTER CONVERSION")
print(clean_df[["Order Date", "Ship Date"]].dtypes)
```

```python
columns = ['Order ID', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name']

for col in columns:
    clean_df[col] = clean_df[col].str.strip().str.title()
    print(f"Text column '{col}' cleaned (strip + title case)")
```

```python
print("Cleaned Data Information:")
clean_df.info()
print()

print("Missing Values after cleaning")
print(clean_df.isnull().sum())
print()

print("Data type of dates in dataframe")
print(clean_df[['Order Date', 'Ship Date']].dtypes)
print()

clean_df = clean_df.drop_duplicates()
print("Duplicate Rows after cleaning")
print(clean_df.duplicated().sum())
print()
```

```python
equal = '=' * 64
print(f"{equal} DATA CLEANING REPORT {equal}\n")
print(f"Rows after cleaning                    : {len(clean_df)}")
print("\nMissing Value Treatment using fillna() and dropna()")
print(f"Missing Before                         : {df.isnull().sum().sum()}")
print(f"Missing After                          : {clean_df.isnull().sum().sum()}")
print("\nDuplicate Value Treatment using drop_duplicates()")
print(f"Duplicates Before                      : {df.duplicated().sum()}")
print(f"Duplicates After                       : {clean_df.duplicated().sum()}")
print("\nDate dtype Treatment using pd.to_datetime()")
print(f"Date type of 'Order Date' Before       : {df['Order Date'].dtypes}")
print(f"Date type of 'Order Date' After        : {clean_df['Order Date'].dtypes}")
print(f"Date type of 'Ship Date' Before        : {df['Ship Date'].dtypes}")
print(f"Date type of 'Ship Date' After         : {clean_df['Ship Date'].dtypes}")
print("\nText cleaning was done to remove extraspace and case sesitive nature was preserved using .str.strip().title()")
print(f'\n1. Missing values were handled\n2. Duplicates were removed\n3. Date was changed to datetime datatype\n4. Text cleaning was performed\n5. DATA READY FOR FEATURE ENGINEERING\n6. Data Lost: {(len(df) - len(clean_df))/len(df)*100}\n')
print('=' * 150)
```

# *Feature Engineering*

```python
featured_data =  clean_df.copy()
```

```python
featured_data.info()
```

```python
featured_data['Order Year'] = featured_data['Order Date'].dt.year
display(featured_data.head())
```

```python
featured_data['Order Month Name'] = featured_data['Order Date'].dt.month_name()
featured_data['Order Month'] = featured_data['Order Date'].dt.month
display(featured_data.head())
```

```python
featured_data['Delivery Days'] = featured_data['Ship Date'] - featured_data['Order Date']
display(featured_data.head())
```

```python
featured_data['Profit Margin (%)'] = (featured_data['Profit'] / featured_data['Sales']) * 100
display(featured_data.head())
```

```python
featured_data['Profit Status'] = np.where(featured_data['Profit']>0, "Profit", "Loss")
display(featured_data.head())
```

```python
featured_data.columns
```

```python
original_cols = df.columns.to_list()
added_cols = featured_data.columns.to_list()
featured_cols = []
for cols in added_cols:
    if cols not in original_cols:
        featured_cols.append(cols)
```

```python
equal = '=' * 60
print(f"{equal} FEATURE ENGINEERING REPORT {equal}\n")
print("New columns created:",featured_cols)
display(featured_data.head())
# 1. Which year has the most orders?
print("Year with highest orders:",featured_data['Order Year'].value_counts().idxmax())
# 2. Which month has the most orders?
print("Month with highest orders:",featured_data['Order Month Name'].value_counts().idxmax())
# 3. What is the average delivery time?
print("Average Delivery Time:", featured_data['Delivery Days'].mean())
# 4. Which orders are loss-making?
print("\nLoss making orders:\n", featured_data[featured_data['Profit Status'] == 'Loss'][['Order ID', 'Product ID', 'Profit', 'Profit Status']])
# 5. What is the average profit margin?
print(f"\nAverage Profit margin: {round(featured_data['Profit Margin (%)'].mean(),3)}%")
print('=' * 150)
```
