# *Data Qulaity Assessment*

```python
print("Total Missing Values in dataset before cleaning:", df.isnull().sum().sum())
missing = df.isnull().sum()
print("\nMissing Values with respect to columns:\n",missing[missing>0])
```

```python
print("Total Duplicate Rows in dataset:", df.duplicated().sum())
print("\nSome duplicate rows in dataset:\n\n")
df[df.duplicated()].head(5)
```

```python
df.describe()
```

```python
print("Minimum Sales values: Rs.",df['Sales'].min())
print("Maximum Sales values: Rs.",df['Sales'].max())
print("Average Profit: Rs.",round(df['Profit'].mean(),2))
print("Maximum Discount: Rs.",df['Discount'].max())

```

```python
print(df['Category'].value_counts())
print("\nCategory with most records:",df['Category'].value_counts().idxmax())
```

```python
print(df['Region'].value_counts())
print("\nRegion with fewest records:",df['Region'].value_counts().idxmin())
```

```python
print(df['Segment'].value_counts())
print("\nSegment appearing most frequently:",df['Segment'].value_counts().idxmax())
```

```python
print("Before conversion:")
print("Data type of Order Date:",df['Order Date'].dtype)
print()
df["Order Date"] = pd.to_datetime(df["Order Date"], format="mixed")
print("After conversion:")
print("Data type of Order Date:",df['Order Date'].dtype)
```

```python
print("Before conversion:")
print("Data type of Ship Date:",df['Ship Date'].dtype)
print()
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="mixed")
print("After conversion:")
print("Data type of Ship Date:",df['Ship Date'].dtype)
```
