## 📖 Concept Explanation

### Why do we explore data?

Imagine you're a store manager. Before making any business decisions, you'd want to know:
- How many transactions happened?
- What products were sold?
- What was the highest and lowest sale?
- Are there any missing entries in the records?

**Data Exploration** helps us answer these basic questions. It's like a health checkup for your data.

### What tools do we use?

We use a Python library called **Pandas**. Pandas gives us special functions (methods) to quickly explore any dataset:

| Method | What It Does |
| :--- | :--- |
| `.head()` | Shows the first 5 rows |
| `.tail()` | Shows the last 5 rows |
| `.shape` | Tells us (rows, columns) count |
| `.columns` | Lists all column names |
| `.dtypes` | Shows the data type of each column |
| `.info()` | Gives a complete summary (types + missing values) |
| `.describe()` | Gives statistics for number columns (mean, min, max, etc.) |
| `.value_counts()` | Counts how many times each value appears |
| `.unique()` | Shows all unique (distinct) values |
| `.nunique()` | Counts how many unique values exist |
| `.isnull().sum()` | Counts missing values in each column |
| `.sort_values()` | Sorts the data by a column |

### What is our dataset about?

We are using a **Sales Data** file (`sales_data.xlsx`) that contains 100 transactions from a retail store. Each row represents one transaction — what was sold, when, at what price, and how much revenue it generated.

## ⚠️ Common Mistakes

Here are the most common mistakes beginners make during data exploration — and how to fix them.

---

### Mistake 1: Using parentheses with `.shape`

````python
# ❌ WRONG — .shape is NOT a method, it's a property
# df.shape()   # This will cause: TypeError: 'tuple' object is not callable

# ✅ CORRECT — use .shape without parentheses
print("Correct way:", df.shape)
````

---

### Mistake 2: Wrong column name (case-sensitive!)

````python
# ❌ WRONG — Python is case-sensitive!
# df['product']   # KeyError: 'product'
# df['PRODUCT']   # KeyError: 'PRODUCT'
# df['revenue']   # KeyError: 'revenue'

# ✅ CORRECT — Use the exact column name as it appears in the dataset
print("Correct:", df['Product'].head(3).tolist())    # Capital 'P'
print("Correct:", df['Revenue'].head(3).tolist())    # Capital 'R'
````

<div style="padding: 12px; border-left: 5px solid #4CAF50; background-color: #e8f5e9; border-radius: 4px; margin: 10px 0;">
<strong>💡 Tip:</strong> If you're unsure of the exact column name, always check with <code>df.columns</code> first!
</div>

---

### Mistake 3: Single brackets vs Double brackets

````python
# ❌ WRONG — Using single brackets to select multiple columns
# df['Product', 'Price']   # KeyError: ('Product', 'Price')

# ✅ CORRECT — Use double brackets (a list inside brackets)
print("Correct way to select multiple columns:")
print(df[['Product', 'Price']].head(3))
````

---

### Mistake 4: Forgetting parentheses with methods

````python
# ❌ WRONG — Forgetting () with methods
# This doesn't CALL the function — it just shows the function object
print("Without ():", df['Revenue'].sum)  # Shows <bound method ...>
print()

# ✅ CORRECT — Add () to actually call the function
print("With ():", df['Revenue'].sum())   # Shows 135800
````

---

### Mistake 5: Using `.describe()` and expecting text columns

````python
# ❌ INCOMPLETE — .describe() by default only shows numeric columns
# It will NOT show Product, Category, etc.
print("Default describe (numeric only):")
print(df.describe())
print()

# ✅ FIX — Use include='all' to see ALL columns including text
print("Describe ALL columns:")
print(df.describe(include='all'))
````

**What happened?**  
- With `include='all'`, we also see stats for text columns:
  - `count` — number of non-null values
  - `unique` — number of unique values
  - `top` — most frequent value
  - `freq` — how many times the most frequent value appears

---

### Mistake 6: Confusing `.count()` with `.shape[0]`

````python
# Both give the same answer HERE because there are no missing values.
# But they work differently!

# .shape[0] — Always gives total number of rows (including empty ones)
print("shape[0]:", df.shape[0])           # Always 100

# .count() — Gives number of NON-MISSING values per column
print("\ncount():")
print(df.count())

# When there are missing values, .count() will show LESS than .shape[0]
# We'll see this in the data cleaning lesson!
````
