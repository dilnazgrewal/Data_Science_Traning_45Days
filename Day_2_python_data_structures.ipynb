## 1. Strings (`str`)
A **string** is a sequence of Unicode characters used for text representation in Python.

### Characteristics:
- **Immutable**: Once created, characters cannot be modified, added, or removed in-place.
- **Ordered & Indexed**: Characters can be accessed using zero-based positive indexing or negative indexing.
- **Allows Duplicates**: The same character can appear multiple times.

### Common String Methods:

| Method | Description | Example |
| :--- | :--- | :--- |
| `.upper()` / `.lower()` | Converts all characters to uppercase / lowercase | `'Python'.lower()` -> `'python'` |
| `.title()` / `.capitalize()` | Capitalizes first letter of words / only the first letter | `'data science'.title()` -> `'Data Science'` |
| `.strip()` / `.lstrip()` / `.rstrip()` | Removes leading and trailing whitespace / only left / only right | `'  text  '.strip()` -> `'text'` |
| `.replace(old, new)` | Replaces occurrences of `old` substring with `new` | `'apple'.replace('p', 'b')` -> `'abble'` |
| `.split(delimiter)` | Splits string into a list of substrings using a delimiter | `'a,b,c'.split(',')` -> `['a', 'b', 'c']` |
| `.join(iterable)` | Joins elements of an iterable of strings with the string as separator | `'-'.join(['a', 'b'])` -> `'a-b'` |
| `.find(sub)` | Returns the lowest index where substring `sub` is found, or `-1` | `'hello'.find('e')` -> `1` |
| `.count(sub)` | Returns the number of non-overlapping occurrences of `sub` | `'banana'.count('a')` -> `3` |
| `.startswith(prefix)` / `.endswith(suffix)` | Returns `True` if string starts with `prefix` / ends with `suffix` | `'hello'.startswith('he')` -> `True` |
| `.isalnum()` / `.isalpha()` / `.isdigit()` | Checks if string contains only alphanumeric / alphabetic / digits | `'123'.isdigit()` -> `True` |

### Built-in Functions with Strings:
- `len(s)`: Returns the number of characters in the string.
- `str(obj)`: Converts an object to its string representation.
- `sorted(s)`: Returns a sorted list of the string's characters.
- `ord(char)` / `chr(code)`: Returns Unicode code point of a character / character of a code point.

## 2. Lists (`list`)
A **list** is a mutable, ordered collection of heterogeneous elements.

### Characteristics:
- **Mutable**: Elements can be modified, added, or removed in-place.
- **Ordered & Indexed**: Maintains insertion order and supports indexing/slicing.
- **Allows Duplicates**: Multiple occurrences of the same element are allowed.
- **Dynamic & Heterogeneous**: Can grow or shrink, and hold elements of different data types simultaneously.

### Common List Methods:

| Method | Description | Example |
| :--- | :--- | :--- |
| `.append(x)` | Appends element `x` to the end of the list | `lst.append(5)` |
| `.extend(iterable)` | Extends list by appending all items from the iterable | `lst.extend([2, 3])` |
| `.insert(index, x)` | Inserts element `x` at the specified index | `lst.insert(0, 9)` |
| `.remove(x)` | Removes the first occurrence of element `x` (raises `ValueError` if not found) | `lst.remove(5)` |
| `.pop(index)` | Removes and returns the element at the index (defaults to the last element) | `item = lst.pop()` |
| `.clear()` | Removes all items from the list | `lst.clear()` |
| `.index(x)` | Returns the zero-based index of the first occurrence of `x` | `idx = lst.index(5)` |
| `.count(x)` | Returns the number of occurrences of `x` | `num = lst.count(5)` |
| `.sort(key, reverse)` | Sorts the list in-place (ascending by default) | `lst.sort()` |
| `.reverse()` | Reverses the list elements in-place | `lst.reverse()` |
| `.copy()` | Returns a shallow copy of the list | `new_lst = lst.copy()` |

### Built-in Functions with Lists:
- `len(lst)`: Returns the number of items.
- `list(iterable)`: Converts an iterable to a list.
- `max(lst)` / `min(lst)`: Returns the largest/smallest element.
- `sum(lst)`: Sums up numeric elements.
- `sorted(lst)`: Returns a new sorted list, leaving original intact.
- `reversed(lst)`: Returns a reverse iterator over the list.
- `enumerate(lst)`: Returns an enumerate object yielding (index, value) tuples.
- `any(lst)` / `all(lst)`: Returns `True` if any/all elements evaluate to `True`.

## 3. Tuples (`tuple`)
A **tuple** is an immutable, ordered sequence of elements, typically used to store collections of heterogeneous data.

### Characteristics:
- **Immutable**: Elements cannot be modified, added, or removed after creation.
- **Ordered & Indexed**: Supports indexing and slicing.
- **Allows Duplicates**: Can contain multiple duplicate elements.
- **Performance & Safety**: Faster than lists and can be used as dictionary keys (since they are hashable).

### Common Tuple Methods:
Since tuples are immutable, they only support two methods:

| Method | Description | Example |
| :--- | :--- | :--- |
| `.count(x)` | Returns the number of occurrences of value `x` | `tup.count(5)` |
| `.index(x)` | Returns the index of the first occurrence of value `x` (raises `ValueError` if not found) | `tup.index(5)` |

### Built-in Functions with Tuples:
- `len(tup)`: Returns number of elements.
- `tuple(iterable)`: Converts an iterable to a tuple.
- `max(tup)` / `min(tup)` / `sum(tup)`: Standard numerical operations.
- `sorted(tup)`: Returns a new *sorted list* containing the tuple's elements.


```python
# --- 1. Tuple Creation, Quirk, and Unpacking ---
print("--- Creation ---")
point = (4, 5)
single_element_tuple = (10,)  # Trailing comma is required for single element
print("point:", point, type(point))
print("Single-element tuple:", single_element_tuple, type(single_element_tuple))

# # Tuple Unpacking / Destructuring
x, y = point
print(f"Unpacked x: {x}, y: {y}")

# # Immutability Check
print("\n--- Immutability Check ---")
# point[0] = 10


# # # --- 2. Tuple Methods ---
# print("\n--- Tuple Methods ---")
my_tuple = (1, 2, 3, 2, 4, 2)
print("Tuple:", my_tuple)
print("count(2):", my_tuple.count(2))
print("index():", my_tuple.index(3))

# --- 3. Built-in Functions on Tuples ---
# print("\n--- Built-in Functions ---")
print("len(my_tuple):", len(my_tuple))
print("max(my_tuple):", max(my_tuple))
print("min(my_tuple):", min(my_tuple))
print("sum(my_tuple):", sum(my_tuple))
print("sorted(my_tuple) (returns a new sorted list):", tuple(sorted(my_tuple)))#SORTED FXN RETURNS A LIST
print("reversed(my_tuple) (returns a new reversed tuple):", tuple(reversed(my_tuple)))

```

## 4. Dictionaries (`dict`)
A **dictionary** is a mutable collection of key-value pairs where keys must be unique and immutable.

### Characteristics:
- **Mutable**: Key-value pairs can be added, updated, or removed.
- **Insertion-Ordered**: Since Python 3.7, dictionaries maintain the order in which keys were inserted.
- **Indexed by Key**: Values are accessed via their unique keys, not index positions.
- **Unique, immutable Keys**: Keys must be of immutable types (strings, numbers, tuples) and cannot be duplicated. Values can be anything.

### Common Dictionary Methods:

| Method | Description | Example |
| :--- | :--- | :--- |
| `.get(key, default)` | Safely retrieves value for `key`; returns `default` if key doesn't exist | `dict.get('age', 0)` |
| `.keys()` | Returns a dynamic view object of all dictionary keys | `dict.keys()` |
| `.values()` | Returns a dynamic view object of all dictionary values | `dict.values()` |
| `.items()` | Returns a dynamic view object of all `(key, value)` tuples | `dict.items()` |
| `.update(other)` | Updates the dictionary with key-value pairs from `other` (dict or iterable) | `dict.update({'gpa': 3.9})` |
| `.pop(key, default)` | Removes the key and returns its value; returns `default` if not found | `dict.pop('age')` |
| `.popitem()` | Removes and returns the last inserted `(key, value)` pair as a tuple | `k, v = dict.popitem()` |
| `.setdefault(key, default)` | If key exists, returns its value. Else, inserts key with value `default` and returns it | `dict.setdefault('status', 'active')` |
| `.clear()` | Removes all key-value pairs | `dict.clear()` |
| `.copy()` | Returns a shallow copy of the dictionary | `new_dict = dict.copy()` |

### Built-in Functions with Dictionaries:
- `len(d)`: Returns the number of key-value pairs.
- `dict(iterable)`: Creates a dictionary from an iterable of key-value pairs.
- `sorted(d)`: Returns a sorted list of the dictionary's *keys*.


```python
# --- 1. Dictionary Creation and Access ---
student = {
    "name": "Praneet",
    "age": 22,
    "courses": ["Python", "Data Science"]
}
print("Original Dict:", student)
# print("Access name directly:", student["name"])

# # Safe access with get()
# print("Access with get() (existing key):", student.get("age",8))
# print("Access with get() (missing key, returns default):", student.get("gpa"))

# # --- 2. Modifying and Adding Entries ---
# print("\n--- Modifying and Adding ---")
student["age"] = 23  # Updates existing key
student["graduated"] = False  # Adds new key-value pair
# print("After modification/addition:", student)

# # Using update()

student2={"name": "Praneet Bawa", "location": "India"}
print(student2)
student2.update(student)
# print("After update():", student)

# # --- 3. Removing Entries ---
# print("\n--- Removing Entries ---")
# removed = student.pop("courses")
# print(f"After pop('courses') (removed {removed}):", student)
# last_item = student.popitem()
# print(f"After popitem() (removed last pair {last_item}):", student)


# # --- 4. Iterating and Dictionary Views ---
# print("\n--- Iterating and Views ---")
print("keys():", list(student.keys()))
print("values():", list(student.values()))
print("items():", list(student.items()))

# print("\nLooping through items:")
for key, val in student.items(): # key,val=("name","Praneet")
    print(f"key in the dictionary: {key}, value in the dictionary: {val}")

# # # --- 5. Dictionary Comprehension ---
# print("\n--- Dictionary Comprehension ---")
# squares = [x**2 for x in range(1, 20)]
# print("Squares dict:", squares)
```

## 5. Sets (`set`)
A **set** is a mutable, unordered collection of unique elements.

### Characteristics:
- **Mutable**: Elements can be added or removed, though set elements themselves must be hashable (immutable).
- **Unordered & Unindexed**: Elements have no defined order, and elements cannot be retrieved by index or sliced.
- **Unique Elements**: Duplicate items are automatically removed upon creation or insertion.
- **Mathematical Operations**: Highly optimized for set math like unions, intersections, and differences.

### Common Set Methods:

| Method | Description | Operator Equivalent |
| :--- | :--- | :--- |
| `.add(x)` | Adds element `x` to the set | N/A |
| `.update(iterable)` | Adds all elements from iterable to the set | N/A |
| `.remove(x)` | Removes element `x` (raises `KeyError` if not found) | N/A |
| `.discard(x)` | Removes element `x` if present (does NOT raise error if missing) | N/A |
| `.pop()` | Removes and returns an arbitrary element | N/A |
| `.union(other)` | Returns the union of two sets | `set1 \| set2` |
| `.intersection(other)` | Returns the intersection of two sets | `set1 & set2` |
| `.difference(other)` | Returns elements in the calling set but not in `other` | `set1 - set2` |
| `.symmetric_difference(other)` | Returns elements in either set, but not both | `set1 ^ set2` |
| `.issubset(other)` | Checks if all elements of this set are in `other` | `set1 <= set2` |
| `.issuperset(other)` | Checks if this set contains all elements of `other` | `set1 >= set2` |
| `.isdisjoint(other)` | Checks if two sets have no common elements | N/A |
| `.clear()` | Removes all elements from the set | N/A |

### Built-in Functions with Sets:
- `len(s)`: Returns the cardinality (number of elements) of the set.
- `set(iterable)`: Converts an iterable to a set (often used for deduplication).
- `max(s)` / `min(s)` / `sum(s)`: Standard numerical operations on elements.
- `sorted(s)`: Returns a new *sorted list* containing the set's elements.


```python
# --- 1. Set Creation, Membership, and Deduplication ---
# print("--- Creation & Membership ---")
colors = {"red", "green", "blue"}
print("Colors set:", colors)

# Check membership (extremely fast O(1) average time complexity)
print("Is 'green' in colors?", "green" in colors)
print("Is 'yellow' in colors?", "yellow" not in colors)

# # # Deduplication helper
# dup_list = [1, 2, 2, 3, 4, 4, 4, 5]
# unique_set = set(dup_list)
# print("Deduplicated list using set():", list(unique_set))

# # --- 2. Modifying Sets ---
# print("\n--- Modifying Sets ---")
# colors.add("yellow")
# print("After add('yellow'):", colors)
colors.update(["green","orange","pink","white"])
# print("After updating:", colors)

# print(colors)
# colors.discard("purple")  # Safe, no error raised
# print("After discard('purple') (no change):", colors)
# colors.remove("purple")  # Will raise KeyError


# # # --- 3. Mathematical Set Operations ---
# print("\n--- Set Operations ---")
setA = {1, 2}
setB = {3, 4, 5, 6}
# print(f"Set A: {setA}")
# print(f"Set B: {setB}")

# print("Union (A | B):", setA.union(setB))
# print("Intersection (A & B):", setA.intersection(setB))
# print("Difference ( B-A):", (setA).difference(setB))
# print("Symmetric Difference (A ^ B):", setB.symmetric_difference(setA))

# # # # --- 4. Subsets and Supersets ---
# print("\n--- Subset & Superset ---")
setC = {3, 4,5}
# print(f"Set C: {setC}")
print("Is C subset of A?", setC.issubset(setA))
print("Is A superset of C?", setA.issuperset(setC))
print("Are A and B disjoint?", setA.isdisjoint(setB))
print(len(setA))
```
