## 📖 Concept Explanation

### What is Data Cleaning?

**Data Cleaning** (also called Data Wrangling) is the process of fixing or removing **incorrect, incomplete, duplicate, or improperly formatted** data in a dataset.

Think of it like this:

| Real World Analogy | Data Cleaning Equivalent |
|---|---|
| Washing vegetables before cooking | Removing errors before analysis |
| Removing rotten fruits from a basket | Dropping corrupt or duplicate rows |
| Filling a blank field in a form | Filling missing values with sensible defaults |
| Correcting a misspelled name | Standardizing text (e.g., fixing case, trimming spaces) |

### Why is Data Cleaning Important?

| Problem | Consequence if NOT cleaned |
|---|---|
| Missing values | Wrong averages, broken calculations |
| Duplicate rows | Sales counted twice → inflated revenue |
| Wrong data types | Dates stored as text → cannot sort by date |
| Inconsistent text | "Electronics" ≠ "electronics" ≠ " Electronics " → treated as 3 different categories |

### Assessment vs. Cleaning

| Data Quality Assessment (Previous Lesson) | Data Cleaning (This Lesson) |
|---|---|
| **Find** the problems | **Fix** the problems |
| `df.isnull().sum()` → "There are 5 missing values" | `df.fillna(value)` → "Fill those 5 missing values" |
| `df.duplicated().sum()` → "There are 3 duplicates" | `df.drop_duplicates()` → "Remove those 3 duplicates" |
| `df.dtypes` → "Date is stored as text" | `pd.to_datetime()` → "Convert text to proper date" |

### The Data Cleaning Workflow

```
Step 1: Load the Data
    ↓
Step 2: Assess the Data (Quick Overview)

    ↓

    ↓

    ↓
✅ Clean Data Ready for Analysis!
```

### Step 2: Handling Missing Values

Missing values appear as **`NaN`** (Not a Number) or **`None`** in Python. They are one of the most common data quality problems.

**Two Main Strategies:**

| Strategy | Method | When to Use |
|---|---|---|
| **Remove** the row | `dropna()` | When very few rows are missing, and they are not important |
| **Fill** with a value | `fillna()` | When you don't want to lose data — fill with mean, median, mode, or a default value |

---

````python
# ---- DETECT MISSING VALUES ----

# .isnull() marks each cell as True (missing) or False (not missing)
# .sum() counts how many True values per column

print("MISSING VALUES COUNT PER COLUMN:")
print("=" * 40)
print(df.isnull().sum())

print(f"\nTOTAL MISSING VALUES: {df.isnull().sum().sum()}")
````

#### 2.3 Strategy 1: Removing Rows with Missing Values — `dropna()`

The `dropna()` function **removes any row that has at least one missing value**.

````python
# ---- STRATEGY 1: DROP ROWS WITH MISSING VALUES ----

# dropna() removes ALL rows that have ANY missing value
# We use a COPY so we don't change the original data yet

df_dropped = df.dropna()  # Returns a NEW DataFrame without missing values

print("BEFORE dropna():")
print(f"  Rows: {len(df)}")

print("\nAFTER dropna():")
print(f"  Rows: {len(df_dropped)}")
print(f"  Rows removed: {len(df) - len(df_dropped)}")
print(f"  Data lost: {((len(df) - len(df_dropped)) / len(df) * 100):.1f}%")
````

**Expected Output:**
```

```

> ⚠️ **Warning:** We lost 16 rows! That's about 14.5% of our data. This might be too aggressive. Let's learn smarter approaches.

---

#### 2.4 Dropping Rows Missing in a SPECIFIC Column — `dropna(subset=)`

Instead of dropping rows missing in **any** column, we can drop only rows missing in a **specific** column.

````python
# ---- DROP ROWS WHERE 'Transaction_ID' IS MISSING ----

# subset=['Transaction_ID'] → Only drop rows where Transaction_ID is NaN
# We can't fill/guess a Transaction ID — it's a unique identifier
# So dropping is the right choice here

df_dropped_id = df.dropna(subset=['Transaction_ID'])

print("DROP only rows where 'Transaction_ID' is missing:")
print(f"  Before: {len(df)} rows")
print(f"  After:  {len(df_dropped_id)} rows")
print(f"  Removed: {len(df) - len(df_dropped_id)} row(s)")
````

**Expected Output:**
```
DROP only rows where 'Transaction_ID' is missing:
  Before: 110 rows
  After:  107 rows
  Removed: 3 row(s)
```

> ✅ **Much better!** We only lost 3 rows (where Transaction_ID was missing) instead of 16 rows.

---

#### 2.5 Strategy 2: Filling Missing Values — `fillna()`

Instead of removing rows, we can **fill** missing values with a replacement. This is usually the better approach because we don't lose data.

**Common fill strategies:**

| Strategy | Code | Best For |
|---|---|---|
| Fill with a **specific value** | `fillna(0)` or `fillna('Unknown')` | When you know a sensible default |
| Fill with the **mean** (average) | `fillna(df['col'].mean())` | Numerical columns with no outliers |
| Fill with the **median** (middle value) | `fillna(df['col'].median())` | Numerical columns WITH outliers |
| Fill with the **mode** (most common) | `fillna(df['col'].mode()[0])` | Categorical (text) columns |

````python
# ---- FILLNA WITH A SPECIFIC VALUE ----

# Let's fill missing 'Category' values with 'Unknown'
# This is useful when we don't know the correct value but want to keep the row

df_fill_example = df.copy()  # Make a copy to not change the original

# BEFORE: Check missing values in Category
print("BEFORE fillna():")
print(f"  Missing Category values: {df_fill_example['Category'].isnull().sum()}")

# Show the rows where Category is missing
print("\n  Rows with missing Category:")
print(df_fill_example[df_fill_example['Category'].isnull()][['Transaction_ID', 'Product', 'Category']])

# FILL: Replace NaN in Category with 'Unknown'
df_fill_example['Category'] = df_fill_example['Category'].fillna('Unknown')

# AFTER: Check again
print("\nAFTER fillna('Unknown'):")
print(f"  Missing Category values: {df_fill_example['Category'].isnull().sum()}")
````

````python
# ---- FILLNA WITH MEAN (for numerical columns) ----

# The MEAN is the average value
# Best for numerical columns when there are no extreme outliers

df_fill_mean = df.copy()

# Calculate the mean of Price column (NaN values are ignored automatically)
price_mean = round(df_fill_mean['Price'].mean(),2)
print(f"Mean of Price: {price_mean}")

# BEFORE: Show rows with missing Price
print(f"\nMissing Price values BEFORE: {df_fill_mean['Price'].isnull().sum()}")
print("\nRows with missing Price:")
print(df_fill_mean[df_fill_mean['Price'].isnull()][['Transaction_ID', 'Product', 'Price']])

# Fill missing Price values with the mean
df_fill_mean['Price'] = df_fill_mean['Price'].fillna(price_mean)

# AFTER
print(f"\nMissing Price values AFTER: {df_fill_mean['Price'].isnull().sum()}")

print(df_fill_mean.loc[66,"Price"])
````

````python
# ---- FILLNA WITH MEDIAN (for numerical columns with outliers) ----

# The MEDIAN is the middle value when data is sorted
# Better than mean when there are extreme values (outliers)

df_fill_median = df.copy()

# Calculate the median of Price column
price_median = df_fill_median['Price'].median()
print(f"Median of Price: {price_median}")
print(f"Mean of Price:   {df_fill_median['Price'].mean():.2f}")
print("→ Median is more robust when data has outliers!")

# Fill missing Price values with the median
df_fill_median['Price'] = df_fill_median['Price'].fillna(price_median)

print(f"\nMissing Price values after fillna(median): {df_fill_median['Price'].isnull().sum()}")
print(df_fill_median.loc[66,"Price"])
````

````python
# ---- FILLNA WITH MODE (for categorical/text columns) ----

# The MODE is the most frequently occurring value
# Best for categorical (text) columns like Product, Category

df_fill_mode = df.copy()

# Calculate the mode (most common value) of Product column
product_mode = df_fill_mode['Product'].mode()[0] # [0] gets the first mode value
print(f"Mode of Product: '{product_mode}'")

# # Show how many times each product appears
print("\nProduct frequency:")
print(df_fill_mode['Product'].value_counts())

# Fill missing Product values with the mode
df_fill_mode['Product'] = df_fill_mode['Product'].fillna(product_mode)

print(f"\nMissing Product values after fillna(mode): {df_fill_mode['Product'].isnull().sum()}")
````

#### 📋 Summary: When to Use Which Strategy?

| Situation | Best Strategy | Our Dataset Example |
|---|---|---|
| Unique identifier missing | `dropna(subset=['col'])` | Drop rows where `Transaction_ID` is missing |
| Cannot guess the correct date | `dropna(subset=['col'])` | Drop rows where `Date` is missing |
| Numerical column, no outliers | `fillna(mean)` | Fill missing `Price` with average price |
| Numerical column, has outliers | `fillna(median)` | Fill missing `Price` with median price |
| Categorical (text) column | `fillna(mode)` | Fill missing `Product` with most common product |
| You know the default value | `fillna('value')` | Fill missing `Category` with `'Unknown'` |

### Step 3: Removing Duplicate Records

A **duplicate row** is an exact copy of another row — every column value is the same.

**Why are duplicates bad?**
- Revenue gets counted twice → wrong total sales
- Product counts get inflated
- Analysis results become unreliable

---

````python
# ---- USING inplace=True ----

# inplace=True modifies the DataFrame DIRECTLY (no need to create a new variable)
# Without inplace=True, you must save the result: df = df.drop_duplicates()

df_example = df.copy()  # Work on a copy

print(f"Before: {len(df_example)} rows")

# Method 1: Without inplace (save the result)
df_example = df_example.drop_duplicates()

# Method 2: With inplace=True (modifies directly)
df_example.drop_duplicates(inplace=True)

print(f"After:  {len(df_example)} rows")
print("\n💡 Both methods do the same thing!")
````

**Expected Output:**
```
Before: 110 rows
After:  100 rows

💡 Both methods do the same thing!
```

> 💡 **Tip:** Both methods do the same thing.
> - `df = df.drop_duplicates()` — saves result to variable
> - `df.drop_duplicates(inplace=True)` — modifies directly

### Step 4: Fixing Data Types

Sometimes data is stored in the **wrong type**. The most common issue is **dates stored as text** (strings).

**Why does it matter?**

| Stored As | What You CAN'T Do |
|---|---|
| Date as text (`"2026-01-15"`) | Cannot sort by date, cannot calculate days between dates |
| Numbers as text (`"150"`) | Cannot calculate sum, average, or do math |

---

#### 4.1 Check Current Data Types

````python
# ---- CHECK DATA TYPES ----

# .dtypes shows the data type of each column
# Common types:
#   int64    → whole numbers (e.g., 150, 2)
#   float64  → decimal numbers (e.g., 3.14, 25.0)
#   object   → text/strings (e.g., 'Laptop', '2026-01-15')
#   datetime → proper date/time values

print("DATA TYPES IN OUR SALES DATASET:")
print("=" * 40)
print(df.dtypes)

print("\n  Notice: 'Date' is stored as 'object' (text), not as a proper date!")
print(f" Sample Date value: '{df['Date'].iloc[0]}' — looks like a date, but is actually text")
````

**Expected Output:**
```

Product            object

    Sample Date value: '2026-01-07' — looks like a date, but is actually text
```

---

#### 4.2 Convert Text to Date — `pd.to_datetime()`

````python
# ---- CONVERT STRING TO DATETIME ----

# pd.to_datetime() converts text dates to proper datetime format
# This allows us to sort by date, extract month/year, calculate date differences

df_typed = df.copy()

# BEFORE conversion
print("BEFORE conversion:")
print(f"  Type of Date column: {df_typed['Date'].dtype}")
print(f"  Sample value: {df_typed['Date'].iloc[0]} (type: {type(df_typed['Date'].iloc[0])})")

# CONVERT: String → Datetime
df_typed['Date'] = pd.to_datetime(df_typed['Date'],format="mixed")

# AFTER conversion
print("\nAFTER conversion:")
print(f"  Type of Date column: {df_typed['Date'].dtype}")
print(f"  Sample value: {df_typed['Date'].iloc[0]} (type: {type(df_typed['Date'].iloc[0])})")
````

**Expected Output:**
```
BEFORE conversion:
  Type of Date column: object
  Sample value: 2026-01-07 (type: <class 'str'>)

AFTER conversion:
  Type of Date column: datetime64[ns]
  Sample value: 2026-01-07 00:00:00 (type: <class 'pandas._libs.tslibs.timestamps.Timestamp'>)
```

> ✅ Now `Date` is a proper datetime! We can sort by date, extract month/year, and more.

---

````python
# ---- WHAT CAN YOU DO WITH DATETIME COLUMNS? ----

# Once converted to datetime, you can extract parts of the date:

# Drop NaN dates first for clean demo
df_date_demo = df_typed.dropna(subset=['Date'])

print(df_date_demo['Date'].head())

# Extract the month name
print("Month names (first 5):\n", df_date_demo['Date'].dt.month_name().head())

# Extract the year
print("Years (first 5):", df_date_demo['Date'].dt.year.head())

# Extract the day of the week
print("Day names (first 5):", df_date_demo['Date'].dt.day_name().head())

# Extract just the day number
print("Day numbers (first 5):", df_date_demo['Date'].dt.day.head())
````

### Step 5: Cleaning Text Columns

Text data often has these problems:

| Problem | Example | Why It's Bad |
|---|---|---|
| Extra spaces | `"  Laptop  "` | `"Laptop"` ≠ `"  Laptop  "` — counted as different products |
| Inconsistent case | `"electronics"` vs `"Electronics"` | Treated as two different categories |
| Typos | `"Elektronics"` | Won't match `"Electronics"` in filters |

**Key Methods:**

| Method | What It Does | Example |
|---|---|---|
| `.str.strip()` | Removes leading/trailing spaces | `"  Laptop  "` → `"Laptop"` |
| `.str.lower()` | Converts all text to lowercase | `"Electronics"` → `"electronics"` |
| `.str.upper()` | Converts all text to UPPERCASE | `"Electronics"` → `"ELECTRONICS"` |
| `.str.title()` | Converts to Title Case | `"electronics"` → `"Electronics"` |
| `.str.replace()` | Replaces specific text | `"Elektronics"` → `"Electronics"` |

### Step 6: Outlier Detection and Handling

**Outliers** are data points that are significantly different from the rest of the data. They can distort statistical analyses and machine learning models.

**How to handle them?**
1. **Detection:** We typically use the **IQR (Interquartile Range)** method or visualization (like boxplots) to find them.
2. **Handling:** We can either **remove** them or **cap** them (replace extreme values with a maximum limit).

---

#### 6.1 Detecting Outliers (IQR Method)

````python
# ---- DETECT OUTLIERS USING IQR ----

# Let's check for outliers in the 'Price' column

# 1. Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1
print(Q1)
print(Q3)
print(IQR)
# 2. Calculate the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("OUTLIER BOUNDARIES:")
print(f"  Lower Bound: {lower_bound}")
print(f"  Upper Bound: {upper_bound}")

# 3. Identify the outliers
outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]
print(f"\n→ Found {len(outliers)} outliers in Price")
if len(outliers) > 0:
    print("\nSample of outliers:")
    print(outliers[['Transaction_ID', 'Product', 'Price']].head())
````

**Expected Output:**
```
OUTLIER BOUNDARIES:
  Lower Bound: -475.0
  Upper Bound: 1125.0

→ Found X outliers in Price
```

---

#### 6.2 Handling Outliers (Capping/Clipping)

````python
# ---- HANDLE OUTLIERS BY CAPPING ----

# Capping (or clipping) replaces values outside the bounds with the bound values.
# This is often better than dropping rows, as we keep the data.

df_no_outliers = df.copy()

print("BEFORE CAPPING:")
print(f"  Max Price: {df_no_outliers['Price'].max()}")

# Apply the .clip() function
df_no_outliers['Price'] = df_no_outliers['Price'].clip(lower=lower_bound, upper=upper_bound)

print("\nAFTER CAPPING:")
print(f"  Max Price: {df_no_outliers['Price'].max()}")

# Verify no outliers remain
remaining_outliers = df_no_outliers[(df_no_outliers['Price'] < lower_bound) | (df_no_outliers['Price'] > upper_bound)]
print(f"\n✅ Remaining outliers: {len(remaining_outliers)}")
````

**Expected Output:**
```
BEFORE CAPPING:
  Max Price: 1000.0

AFTER CAPPING:
  Max Price: 1000.0

✅ Remaining outliers: 0
```

## ⚠️ Common Mistakes

Here are the most common beginner mistakes and how to fix them.

---

### ❌ Mistake 1: Forgetting to Save the Result

````python
# ---- MISTAKE 1: FORGETTING TO SAVE THE RESULT ----

# Reload the messy data for this demo
df_test = pd.read_excel('sales_data_messy.xlsx')

# ❌ WRONG: This does NOT change df_test!
df_test.drop_duplicates()  # This returns a new DataFrame, but we don't save it!
print(f"❌ WRONG — Rows after 'cleaning': {len(df_test)}")  # Still has duplicates!

# ✅ FIX Option 1: Save the result back to the variable
df_test = pd.read_excel('sales_data_messy.xlsx')
df_test = df_test.drop_duplicates()  # Save the result!
print(f"✅ FIX 1 — Rows after cleaning: {len(df_test)}")

# ✅ FIX Option 2: Use inplace=True
df_test2 = pd.read_excel('sales_data_messy.xlsx')
df_test2.drop_duplicates(inplace=True)  # Modifies directly!
print(f"✅ FIX 2 — Rows after cleaning: {len(df_test2)}")
````

**Expected Output:**
```
❌ WRONG — Rows after 'cleaning': 110
✅ FIX 1 — Rows after cleaning: 100
✅ FIX 2 — Rows after cleaning: 100
```

> ⚠️ Always either: `df = df.drop_duplicates()` OR `df.drop_duplicates(inplace=True)`

---

### ❌ Mistake 2: Using `dropna()` Without Thinking

````python
# ---- MISTAKE 2: USING dropna() TOO AGGRESSIVELY ----

df_test3 = pd.read_excel('sales_data_messy.xlsx')

# ❌ WRONG: Dropping ALL rows with ANY missing value — loses too much data!
df_wrong = df_test3.dropna()
print(f"❌ WRONG — dropna() removed {len(df_test3) - len(df_wrong)} out of {len(df_test3)} rows!")
print(f"   Only {len(df_wrong)} rows remaining — we lost {((len(df_test3) - len(df_wrong)) / len(df_test3) * 100):.0f}% of data!")

# ✅ FIX: Use dropna(subset=) for IDs and dates, and fillna() for others
df_smart = df_test3.copy()
df_smart = df_smart.dropna(subset=['Transaction_ID', 'Date'])  # Only drop critical missing
df_smart['Product'] = df_smart['Product'].fillna(df_smart['Product'].mode()[0])   # Fill Product
df_smart['Category'] = df_smart['Category'].fillna(df_smart['Category'].mode()[0]) # Fill Category
df_smart['Price'] = df_smart['Price'].fillna(df_smart['Price'].median())           # Fill Price
print(f"\n✅ FIX — Smart cleaning: {len(df_smart)} rows remaining")
print(f"   Missing values: {df_smart.isnull().sum().sum()}")
````

**Expected Output:**
```
❌ WRONG — dropna() removed 16 out of 110 rows!
   Only 94 rows remaining — we lost 15% of data!

✅ FIX — Smart cleaning: 104 rows remaining
   Missing values: 0
```

> ⚠️ Think carefully about WHICH rows to drop and WHICH to fill! Smart cleaning saved 10 extra rows.

---

### ❌ Mistake 3: Removing Duplicates BEFORE Cleaning Text

````python
# ---- MISTAKE 3: WRONG ORDER OF OPERATIONS ----

# Imagine our Product column had text issues like '  Laptop  ' and 'Laptop'
# If we remove duplicates FIRST, they won't be detected as duplicates!

# Let's demonstrate with a small example:
demo = pd.DataFrame({
    'Product': ['  Laptop  ', 'Laptop', 'Tablet', '  Laptop  '],
    'Price': [1000, 1000, 400, 1000]
})

# ❌ WRONG: Remove duplicates FIRST, then clean text
demo_wrong = demo.drop_duplicates()
demo_wrong['Product'] = demo_wrong['Product'].str.strip()
print(f"❌ WRONG ORDER — Rows: {len(demo_wrong)}, Duplicates left: {demo_wrong.duplicated().sum()}")

# ✅ CORRECT: Clean text FIRST, then remove duplicates
demo_right = demo.copy()
demo_right['Product'] = demo_right['Product'].str.strip()  # Clean text FIRST
demo_right = demo_right.drop_duplicates()                   # THEN remove duplicates
print(f"✅ RIGHT ORDER — Rows: {len(demo_right)}, Duplicates left: {demo_right.duplicated().sum()}")
````

**Expected Output:**
```
❌ WRONG ORDER — Rows: 3, Duplicates left: 1
✅ RIGHT ORDER — Rows: 2, Duplicates left: 0
```

> ⚠️ **Remember the correct order:**
> 1. Clean text (strip + title) ← **FIRST**
> 2. Handle missing values
> 3. Remove duplicates ← **AFTER text is clean**
> 4. Fix data types

### ❌ Mistake 4: Wrong Syntax for `fillna()` with Mode

````python
# ---- MISTAKE 4: WRONG MODE SYNTAX ----

df_test5 = pd.read_excel('sales_data_messy.xlsx')

# ❌ WRONG: mode() returns a Series, not a single value!
# df_test5['Product'].fillna(df_test5['Product'].mode())  # This won't work as expected!
print("❌ WRONG: df['Product'].mode() returns:")
print(df_test5['Product'].mode())  # Returns a Series with index
print("   This is a Series, not a single value!\n")

# ✅ FIX: Use mode()[0] to get the FIRST (most common) value
print("✅ FIX: df['Product'].mode()[0] returns:")
print(df_test5['Product'].mode()[0])  # Returns just the value
print("   This is a single value — perfect for fillna()!")
````
